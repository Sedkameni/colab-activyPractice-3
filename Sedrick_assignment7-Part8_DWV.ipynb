```python
#### Part 8 - Regular Expression

######### -  Compile a regular expression pattern to split the text "one1two2three3four4" into words.

import re

text = "one1two2three3four4"
pattern = r"\d+"
new_text = re.split(pattern, text)
print(new_text)
```

```python
#### - Use the findall function to find all word patterns in the string "My phone number is 123-456-7890".
text1 = "My phone number is 123-456-7890"
pattern1 = r'\w+'
phone_number = re.findall(pattern1, text1)
print(phone_number)
```

```python
### - Use the search function to find the first occurrence of the pattern "number" in the string "Contact number: 9876543210".
#search_result = re.search(r'number', "Contact number: 9876543210")
#print(search_result.group())
text2 = "Contact number: 9876543210"
pattern2 = r"number"
result = re.search(pattern2, text2)
print(result)
```

```python
#### - Use the match function to check if the string "Python" starts with the pattern "Py".
text3 = "Python"
pattern3 = r"^Py"
result = re.match(pattern3, text3)
print(bool(result))
```
