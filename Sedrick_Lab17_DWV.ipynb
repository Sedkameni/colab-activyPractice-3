```python
### - Task 1
### 1.1 - Import libraries. Load and explore the "heartDiseaseProcessedCleveland.csv" dataset available at the bottom of this page.
import pandas as pd

# Load the dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data"
column_names = ["age", "sex", "cp", "trestbps", "chol", "fbs", "restecg", "thalach", "exang", "oldpeak", "slope", "ca", "thal", "target"]
heart_data = pd.read_csv(url, header=None, names=column_names, na_values="?")

# Display first few rows
print(heart_data.head())

# Describe the dataset
print(heart_data.describe())
```

```python
### 1.2 - Updating Figures and Layouts
import plotly.graph_objects as go
fig = go.Figure()
```

```python
### 1.3 - Exercise 1: Create a scatter plot to visualize the relationship between age and cholesterol levels using Plotly.Graph_Objects.
### Customize the plot using various parameters. Update the layout with template, titles and axis labels.
# Adding traces
fig.add_trace(go.Scatter(x=heart_data['age'], y=heart_data['chol'], mode='markers', name='Cholesterol'))

# Updating layout
fig.update_layout(title='Cholesterol Levels by Age',
                  xaxis_title='Age',
                  yaxis_title='Cholesterol',
                  template='plotly_dark')

fig.show()
```

```python
### 1.4 - Updating Traces and Axes: Update the traces to customize marker properties (e.g., size, color, line width).
####      Update the x and y axes with custom font properties (e.g., size, family, color).
# Updating traces
fig.update_traces(marker=dict(size=10, color='rgba(152, 0, 0, .8)', line=dict(width=2, color='DarkSlateGrey')))

# Updating axes
fig.update_xaxes(title_font=dict(size=18, family='Courier', color='crimson'),
                 tickfont=dict(size=14, family='Arial', color='crimson'))
fig.update_yaxes(title_font=dict(size=18, family='Courier', color='crimson'),
                 tickfont=dict(size=14, family='Arial', color='crimson'))

fig.show()
```

```python
#### 1.5 - Subplots with make_subplots: Create a subplot with two plots side-by-side using make_subplots.
########    Add traces to each subplot (e.g., scatter plots of age vs cholesterol, and age vs max heart rate).
####        Update the layout with template, title and axis labels.
from plotly.subplots import make_subplots

# Creating subplots
subfig = make_subplots(rows=1, cols=2, subplot_titles=('Age vs Cholesterol', 'Age vs Max Heart Rate'))

# Adding traces
subfig.add_trace(go.Scatter(x=heart_data['age'], y=heart_data['chol'], mode='markers', name='Cholesterol'), row=1, col=1)
subfig.add_trace(go.Scatter(x=heart_data['age'], y=heart_data['thalach'], mode='markers', name='Max Heart Rate'), row=1, col=2)

# Updating layout
subfig.update_layout(title='Heart Disease Data',
                  xaxis_title='Age',
                  yaxis_title='Value',
                  template='plotly_dark')

subfig.show()
```

```python
#### 1.6 - Figure Factories: Create a dendrogram using the first age and chol columns of the dataset with Figure_Factories.
import plotly.figure_factory as ff

# Dendrogram

dendrofig = ff.create_dendrogram(heart_data[['age','chol']], orientation= 'left')
dendrofig.update_layout(title='Dendrogram of Heart Disease Data',
                  xaxis_title='Chol',
                  yaxis_title='Age')
dendrofig.show()
```

```python
#### 1.7 - Create a quiver plot with age and max heart rate as coordinates and trestbps and cholesterol as vectors.
# Quiver plot
x = heart_data['age']
y = heart_data['thalach']
u = heart_data['trestbps']
v = heart_data['chol']
quiver = ff.create_quiver(x, y, u, v)
quiver.update_layout(title='Quiver Plot of Heart Disease Data',
                  xaxis_title='Age',
                  yaxis_title='Max Heart Rate')
quiver.show()
```

```python
### 1.8 - Plotly.io: Create a scatter plot of age vs max heart rate, customize it, and save it as an image file.
import plotly.io as pio
import kaleido
# Create the figure
scatterfig = go.Figure()

# Add scatter trace
scatterfig.add_trace(go.Scatter(
    x=heart_data['age'],
    y=heart_data['thalach'],
    mode='markers',
    marker=dict(size=10, color='rgba(152, 0, 0, .8)', line=dict(width=2, color='DarkSlateGrey'))
))

# Update layout
scatterfig.update_layout(
    title='Max Heart Rate by Age',
    xaxis_title='Age',
    yaxis_title='Max Heart Rate',
    template='plotly_white'
)

# Save the figure as an image
pio.write_image(scatterfig, 'max_heart_rate_by_age.png')

# Show the figure
scatterfig.show()
```

```python
### 1.9 - Create a bar plot of chest pain type distribution, save it to a JSON file, and read it back from the JSON file.
# Create the figure
barfig = go.Figure()

# Add bar trace
barfig.add_trace(go.Bar(
    x=heart_data['cp'].value_counts().index,
    y=heart_data['cp'].value_counts().values,
    marker=dict(color='blue')
))

# Update layout
barfig.update_layout(
    title='Distribution of Chest Pain Type',
    xaxis_title='Chest Pain Type',
    yaxis_title='Count',
    template='plotly_white'
)

# Save the figure as a JSON file
pio.write_json(barfig, 'chest_pain_distribution.json')

# Read the figure from the JSON file
fig_from_json = pio.read_json('chest_pain_distribution.json')

# Show the figure
fig_from_json.show()
```

```python
#### 1.10 - Create a box plot of cholesterol levels by sex, customize it, and save it as an HTML file on auto open.
# Create the figure
boxfig = go.Figure()

# Add box trace
boxfig.add_trace(go.Box(
    x=heart_data['sex'],
    y=heart_data['chol'],
    boxpoints='all',
    jitter=0.3,
    pointpos=-1.8,
    marker=dict(color='green')
))

# Update layout
boxfig.update_layout(
    title='Cholesterol Levels by Sex',
    xaxis_title='Sex',
    yaxis_title='Cholesterol',
    template='plotly_white',
    xaxis=dict(tickmode='array', tickvals=[0, 1], ticktext=['Female', 'Male'])
)

# Save the figure as an HTML file
pio.write_html(boxfig, 'cholesterol_by_sex.html', auto_open=True)

# Show the figure
boxfig.show()
```

```python
### Task 2:
### 2.1 - Install dash package: pip install dash. Import libraries
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
```

```python
#### 2.2 - Load the "gapminderDataFiveYear.csv" dataset.
# Load the dataset
df = px.data.gapminder()
```

```python
#### 2.3 - Initialize the Dash app and define the layout.
# Initialize the Dash app
app = dash.Dash(__name__)

# Define the layout
app.layout = html.Div([
    html.H1("Gapminder Data Dashboard"),
    html.Label('Select Year:'),
    dcc.Slider(
        id='year-slider',
        min=df['year'].min(),
        max=df['year'].max(),
        value=df['year'].min(),
        marks={str(year): str(year) for year in df['year'].unique()},
        step=None
    ),
    dcc.Graph(id='life-exp-vs-gdp'),
    html.Label('Select Continent:'),
    dcc.Dropdown(
        id='continent-dropdown',
        options=[{'label': c, 'value': c} for c in df['continent'].unique()],
        value='Asia',
        clearable=False
    ),
    dcc.Graph(id='population-vs-gdp')
])
```

```python
### 2.4 - Define the callback for updating the life expectancy vs GDP graph.
# Define the callback for updating the life expectancy vs GDP graph
@app.callback(
    Output('life-exp-vs-gdp', 'figure'),
    [Input('year-slider', 'value')]
)
def update_life_exp_vs_gdp(selected_year):
    filtered_df = df[df['year'] == selected_year]
    fig = px.scatter(filtered_df, x='gdpPercap', y='lifeExp',
                     size='pop', color='continent',
                     hover_name='country', log_x=True, size_max=60,
                     title=f'Life Expectancy vs GDP per Capita in {selected_year}')
    return fig
```

```python
#### 2.5 - Define the callback for updating the population vs GDP graph.
# Define the callback for updating the population vs GDP graph
@app.callback(
    Output('population-vs-gdp', 'figure'),
    [Input('continent-dropdown', 'value')]
)
def update_population_vs_gdp(selected_continent):
    filtered_df = df[df['continent'] == selected_continent]
    fig = px.scatter(filtered_df, x='gdpPercap', y='pop',
                     size='pop', color='year',
                     hover_name='country', log_x=True, size_max=60,
                     title=f'Population vs GDP per Capita in {selected_continent}')
    return fig
```

```python
### 2.6 - Run the app.
# Run the app
if __name__ == '__main__':
    app.run(debug=True)
```
