```python
### Task 2 : Work with SQLite3.
## import libraries
import pandas as pd
import sqlite3

### Create a connection. Create a cursor object.

# Create a connection
conn = sqlite3.connect('Store.db')

# Create a cursor object
cur = conn.cursor()
```

```python
### 2 - Load tables: Product and Aisle
# Load Product table into a DataFrame
Product = pd.read_sql('SELECT * FROM Product', conn)

# Load Aisle table into a DataFrame
Aisle = pd.read_sql('SELECT * FROM aisle', conn)

# Print dataframe
print("\nWe display the Product table:\n")
print(Product)
print("\nWe display the Aisle table:\n")
print(Aisle)
```

```python
#### 3 - Join tables:
#‘Product’ and ‘Aisle’ using Aisle_Id column in both tables.
#Select from Product table (columns – Product_Id, Product_Name , Price) and
#from Aisle table (columns – Aisle_Id, Aisle_Name) in the resulting query.

sql = '''
SELECT p.Product_Id,
       p.Product_Name,
       p.Price,
       p.Aisle_Id,
       a.Aisle_Id,
       a.Aisle_Name
FROM Product p
INNER JOIN Aisle a
ON p.Aisle_Id = a.Aisle_Id;
'''
```

```python
### - Query database to get query and load into a Python object.
#Executing the query
cur.execute(sql)


#Fetching 1st row from the table
results = cur.fetchall()
```

```python
#### 2. 4 - Saving query result into a DataFrame.
columns = ['Product_Id', 'Product_Name', 'Price', 'Product_Aisle_Id', 'Aisle_Id', 'Aisle_Name']
df = pd.DataFrame(results, columns=columns)
print(df)
```

```python
#### 2.5 - Export DataFrame to a CSV file named “Product_Aisle.csv”
df.to_csv('Product_Aisle.csv', index=False)
```

```python
#### - close connection
conn.close()
```
