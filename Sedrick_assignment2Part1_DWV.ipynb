```python
#Part 1 :
#### PART A : Load from and read to EXCEL file
#####  1 - Import pandas library.
import pandas as pd

##### 2 - Load the file: sales.xlsx into a DataFrame.
file=pd.ExcelFile("sales.xlsx")
```

```python
##### 3 - Print all the sheet names.
print(file.sheet_names)
```

```python
##### 4 - Read into a new DataFrame using parse the sheet name – ‘Phone&Electronics’
#with these specifications: Skip first 10 rows, show the next 15 rows, and skip the third column. Then print.
sheet1=file.parse(sheet_name='Phone&Electronics', skiprows=15, nrows=15, usecols=[0,1,3,4,5,6,7])
print(sheet1)
```

```python
##### PART B: Using read_excel
###  Reload the Sales file with sheet name – ‘Office Supply’
## into a new DataFrame with these specifications: Skip rows 2, 15, 28, 36 and 41, and skip the second column. Then print.
rows_to_skip = [1, 14, 27, 35, 40]
df = pd.read_excel('sales.xlsx', sheet_name='Office Supply', skiprows=[1, 14, 27, 35, 40], usecols=[0,2,3,4,5,6])
print(df)
```

```python
####### PART C: Using ExcelWriter and to_excel
###### Save to a new excel file named ‘sales_sample.xlsx’ the following:
####### Output of PART A to a sheet called ‘Sample1’.
####### Output of PART B to a sheet called ‘Sample2’.

with pd.ExcelWriter('sales_sample.xlsx') as writer:
    sheet1.to_excel(writer, sheet_name='Sample1')
    df.to_excel(writer, sheet_name='Sample2')
```
