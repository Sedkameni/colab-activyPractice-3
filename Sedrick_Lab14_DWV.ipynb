```python
### Task 1
### 1.1 Import libraries. Load and explore the "insurance.csv" dataset available at the bottom of this page.
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load Insurance dataset
url = 'https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/insurance.csv'
insurance = pd.read_csv(url)

# Display the first few rows of the dataset
print(insurance.head())
```

```python
### 1.2 - Setting plots to have a white grid.
sns.set_style('whitegrid')
```

```python
#### 1.3 - Catplot: Create a strip plot to show the relationship between smoker and charges. Explore jitter, palette, dodge parameters.
#  Catplot: using kind='strip'
sns.catplot(data=insurance, x='smoker', y='charges', kind='strip', jitter=True, palette='Set1', dodge=True)
plt.title('Strip Plot of Smoker vs Charges')
plt.xlabel('Smoker')
plt.ylabel('Charges')
plt.show()
```

```python
#### 1.4 - Create a swarm plot to show the relationship between smoker and charges. Explore palette, dodge parameters.
# Catplot: using kind='swarm'
sns.catplot(data=insurance, x='smoker', y='charges', kind='swarm', palette='Set2', dodge=True)
plt.title('Swarm Plot of Smoker vs Charges')
plt.xlabel('Smoker')
plt.ylabel('Charges')
plt.show()
```

```python
#### 1.5 - Create a box plot to show the relationship between region and charges. Explore palette, hue parameters.
## -  Catplot: using kind='box'
sns.catplot(data=insurance, x='region', y='charges', kind='box', palette='Set3', hue='smoker')
plt.title('Box Plot of Region vs Charges')
plt.xlabel('Region')
plt.ylabel('Charges')
plt.show()
```

```python
### 1.6 - Create a violin plot to show the relationship between region and charges. Explore palette, hue, split parameters.
#  Catplot: using kind='violin'
sns.catplot(data=insurance, x='region', y='charges', kind='violin', palette='muted', hue='smoker', split=True)
plt.title('Violin Plot of Region vs Charges')
plt.xlabel('Region')
plt.ylabel('Charges')
plt.show()
```

```python
#### 1.7 - Create a boxen plot to show the relationship between region and charges. Explore palette, hue parameters.
#  Catplot: using kind='boxen'
sns.catplot(data=insurance, x='region', y='charges', kind='boxen', palette='pastel', hue='smoker')
plt.title('Boxen Plot of Region vs Charges')
plt.xlabel('Region')
plt.ylabel('Charges')
plt.show()
```

```python
#### 1.8 - Create a point plot to show the relationship between smoker and charges. Explore palette, errorbar, markers, linestyles parameters.
#  Catplot: using kind='point'
sns.catplot(data=insurance, x='region', y='charges', kind='point', palette='dark', hue='smoker', errorbar='sd', markers=["o", "x"], linestyles=["-.", "--"])
plt.title('Point Plot of Smoker vs Charges')
plt.xlabel('Smoker')
plt.ylabel('Charges')
plt.show()
```

```python
#### 1.9 - Create a bar plot to show the relationship between sex and charges. Explore palette, hue, errorbar parameters.
# Catplot: using kind='bar'
sns.catplot(data=insurance, x='sex', y='charges', kind='bar', palette='muted', hue='smoker', errorbar='sd')
plt.title('Bar Plot of Sex vs Charges')
plt.xlabel('Sex')
plt.ylabel('Charges')
plt.show()
```

```python
#### 1.10 - Create a count plot to show the number of occurrences for each region. Explore palette, hue parameters.
## Catplot: using kind='count'
sns.catplot(data=insurance, x='region', kind='count', palette='Set2', hue='smoker')
plt.title('Count Plot of Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()
```

```python
#### Task 2:

##### 2.1- Import libraries. Load and explore the "insurance.csv" dataset.
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load Insurance dataset
url = 'https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/insurance.csv'
insurance = pd.read_csv(url)

# Display the first few rows of the dataset
print(insurance.head())

##### 2.2 - Setting plots to have a dark grid.
sns.set_style('darkgrid')
```

```python
#### 2.3 - Stripplot: Using stripplot, show the relationship between charges and smoker. Explore sex as hue, jitter, palette, and dodge parameters.
# Strip plot
sns.stripplot(data=insurance, x='charges', y='smoker', hue='sex', jitter=True, palette='Set2', dodge=True, marker='^')
plt.title('Strip Plot')
plt.legend(title='Sex',loc=4)
plt.show()
```

```python
#### 2.4 - Swarmplot: Using swarmplot, show the relationship between smoker and charges. Explore region as hue, palette, dodge, and size parameters.
# Swarm plot
sns.swarmplot(data=insurance, x='smoker', y='charges', palette='magma', size=5, hue='region', dodge=True)
plt.title('Swarm Plot of Smoker vs Charges')
plt.xlabel('Smoker')
plt.ylabel('Charges')
plt.legend(title='Region')
plt.show()
```

```python
#### 2.5 - Boxplot: Using boxplot, show the relationship between region and charges. Explore sex as hue, size, notch, width, and palette parameters.
# Box plot
sns.boxplot(data=insurance, x='region', y='charges', palette='Set1', hue='sex', notch=True, width=0.7)
plt.title('Box Plot of Region vs Charges')
plt.xlabel('Region')
plt.ylabel('Charges')
plt.legend(title='Sex')
plt.show()
```

```python
##### 2.6 - Violin plot: Create a violin plot to show the relationship between charges and sex. Explore smoker as hue, palette, split, and scale parameters.
# Violin plot
sns.violinplot(data=insurance, x='charges', y='sex', palette='cool', hue='smoker', split=True, scale='width')
plt.title('Violin Plot of Charges vs Sex')
plt.xlabel('Charges')
plt.ylabel('Sex')
plt.legend(title='Smoker')
plt.show()
```

```python
#### 2.7 - Boxen plot: Using boxenplot, show the relationship between region and charges. Explore smoker as hue, palette, and k_depth parameters.
# Boxen plot
sns.boxenplot(data=insurance, x='region', y='bmi', palette='viridis', hue='smoker', k_depth='full')
plt.title('Boxen Plot of Region vs BMI')
plt.xlabel('Region')
plt.ylabel('BMI')
plt.legend(title='Smoker', loc=9)
plt.show()
```

```python
#### 2.8 - Point plot: Using pointplot, show the relationship between smoker and charges. Explore sex as hue, palette, join, markers, and linestyles parameters.
# Point plot
sns.pointplot(data=insurance, x='region', y='charges', palette='bright', hue='sex', join=True, markers=['>', 'o'], linestyles=['-','-.'])
plt.title('Point Plot of Region vs Charges')
plt.xlabel('Region')
plt.ylabel('Charges')
plt.legend(title='Sex')
plt.show()
```

```python
#### 2.9 - Bar plot: Using barplot, show the relationship between sex and charges. Explore smoker as hue, palette, errorbar, and capsize parameters.
# Bar plot
sns.barplot(data=insurance, x='sex', y='charges', palette='Set3', hue='smoker', errorbar='ci', capsize=0.1)
plt.title('Bar Plot of Sex vs Charges')
plt.xlabel('Sex')
plt.ylabel('Charges')
plt.legend(title='Smoker')
plt.show()
```

```python
#### 2.10 - Count plot: Using countplot, show the number of occurrences for each region. Explore smoker as hue, palette, hue parameters.
# Count plot
sns.countplot(data=insurance, x='sex', palette='Set1', hue='smoker')
plt.title('Count Plot of Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(title='Smoker')
plt.show()
```
