```python
### Part 1 - Copy these lines of code into your Python file
### - Load the Airbnb New York City Listings dataset using Pandas.
import pandas as pd
df = pd.read_csv("AB_NYC_2019.csv")

###Display the first few rows and describe its basic statistics.
print("\nWe display the first few rows here:\n", df.head())
print("\nWe display the basic statistics here:\n", df.describe)
```

```python
#### Part 2 - Data Reshaping in Pandas:
### 2.1 - Pivoting: Create a pivot table showing the average price for each room type in each neighborhood group.
pivot_table = df.pivot_table(values='price', index='room_type', columns='neighbourhood', aggfunc='mean')
print(pivot_table)
```

```python
#### 2.2 - Melting: Melt the pivot table back to long format.
melted_table = pivot_table.reset_index().melt(id_vars='room_type', value_vars=pivot_table.columns)
print(melted_table)
```

```python
### 2.3 - Stacking: Stack the dataframe and convert it to a multi-index series.
stacked_df = pivot_table.stack()
print(stacked_df)
```

```python
### 2.4 - Unstacking: Unstack the multi-index series back to a dataframe.
unstacked_df = stacked_df.unstack()
print(unstacked_df)
```

```python
### Part 3 - Categorical Extension Type in Pandas
## 3.1 - Convert the room_type column to a categorical data type using the Categorical constructor.
df['room_type'] = pd.Categorical(df['room_type'])
print(df['room_type'].dtype)
```

```python
 ### 3.2 - Use the astype('category') method to convert the neighbourhood_group column to a categorical data type.
 df['neighbourhood_group'] = df['neighbourhood_group'].astype('category')
 print(df['neighbourhood_group'].dtype)
```

```python
### Part 4 - Using the cat Attribute
###### 4.1 -  Use the cat.codes attribute to view the category codes for the room_type column.
print("The category codes for room type column is:")
print(df["room_type"].cat.codes)
```

```python
### 4.2 - Use the cat.categories attribute to view the categories of the neighbourhood_group column.
print("The categories for neighbourhood group column is:")
print(df["neighbourhood_group"].cat.categories)
```

```python
### 4.3 - Add a new category to the room_type column using cat.add_categories.
df["room_type"] = df["room_type"].cat.add_categories("Shared Room")
print(df["room_type"].cat.categories)
```

```python
#### 4.4 - Remove unused categories from the room_type column using cat.remove_unused_categories.
df["room_type"] = df["room_type"].cat.remove_unused_categories()
print(df["room_type"].cat.categories)
```

```python
#### 4.5 - Set new categories for the neighbourhood_group column using cat.set_categories.
new_categories = ['Bronx', 'Brooklyn', 'Manhattan', 'Queens', 'Staten Island', 'Other']
df["neighbourhood_group"] = df["neighbourhood_group"].cat.set_categories(new_categories)
print(df["neighbourhood_group"].cat.categories)
```

```python
##### Part 5 - Computations with Categoricals
##### 5.1 - Use the value_counts method to count the occurrences of each room type.
print("\nThen number of occurence of each room type is:\n", df["room_type"].value_counts())
```

```python
##### 5.2 - Find the mode of the room_type column.
print("\nThe mode of room type is:\n", df["room_type"].mode()[0])
```

```python
#### 5.3 - Use the sort_values method to sort the dataframe by the neighbourhood_group column.
print("\nThe dataframe sorted by neighbourhood group is:\n", df.sort_values(by="neighbourhood_group"))
```
