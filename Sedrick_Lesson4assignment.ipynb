```python
# Task 9: Iterators and Generators
# Create an iterator for the first 5 even numbers.
# Then, create a generator for the same.

# Your code here:
# Iterator for the first 5 even numbers
class EvenIterator:
    def __init__(self):
        self.num = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.num < 10:
            even = self.num
            self.num += 2
            return even
        else:
            raise StopIteration

# Test iterator
print("\n Here is the result for iterator\n")

for n in EvenIterator():
    print(n)


# Generator for the first 5 even numbers
def even_generator():
    for n in range(0, 10, 2):
        yield n

# Test generator
print("\n Here is the result for Generator\n")
for n in even_generator():
    print(n)
```

```python
# Task 10: Map, Reduce, and Filter
# Use map to square all numbers in a list.
# Use reduce to find the product of all numbers in a list.
# Use filter to get only even numbers from a list.

numbers = [1, 2, 3, 4, 5]

# Your code here:
from functools import reduce

def multi_num(a,b):
    return a*b


print("Map:", list(map(lambda x:x**2, numbers)))  # map() applies a function to each element in an iterable (like a list).
print("Filter:", list(filter(lambda x:x%2==0, numbers)))
print("Reduce:", reduce(multi_num, numbers))
```

```python
# Task 11: Object-Oriented Programming - Creating a Class
# Create a class called 'Rectangle' with attributes 'length' and 'width'.
# Include methods to calculate area and perimeter.

# Your code here:
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def perimeter(self):
        return 2*(self.length + self.width)

    def area(self):
        return self.length * self.width

    def __str__(self):
        return f"length {self.length} and width {self.width}"

alec = Rectangle(35, 15)
print("This rectangle has", alec)
print("The perimeter is", alec.perimeter())
print("The area is", alec.area())
```

```python
# Python and Pandas Assessment Notebook

# Task 12: Pandas Data Analysis
# Using the following DataFrame, perform these tasks:
# 1. Find the average salary by department
# 2. Get the names of employees with salary > 60000
# 3. Add a new column 'Bonus' which is 10% of salary

df_employees = pd.DataFrame({
    'Name': ['John', 'Jane', 'Bob', 'Alice', 'Charlie'],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR'],
    'Salary': [55000, 65000, 70000, 60000, 58000]
})

# Your code here:
# 1 - Find the average salary by department
avg_salary = df_employees.groupby('Department')['Salary'].mean()
print("\nAverage salary by department:\n", avg_salary)

#2 - Get the names of employees with salary > 60000
print("\nEmployees with salary > 60000:\n", df_employees[df_employees['Salary'] > 60000]['Name'])

#3 - Add a new column 'Bonus' which is 10% of salary
df_employees['Bonus'] = df_employees['Salary'] * 0.1
print("\nNew DataFrame with Bonus column:\n", df_employees)
```
