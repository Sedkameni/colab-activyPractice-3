2 - Data Extraction and Cleaning:

*   Handle missing values in the `Customer_Age` and 'Customer_Location' columns by filling them with the average age and a placeholder 'Unknown' respectively.





```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('assignment_sales_data.csv')
# Fill missing values
df.fillna({'Customer_Location': 'Unknown', 'Customer_Age': df['Customer_Age'].mean()}, inplace=True)
# Correct formatting issues (example: standardize customer location names)
df['Customer_Location'] = df['Customer_Location'].str.title()
print(df)
```

*   Calculate total sales.
*   Calculate the average customer age

```python
import csv
total_sales = 0
total_age = 0
customer_count = 0
with open('assignment_sales_data.csv', mode='r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)  # Skip the header row
    for row in csv_reader:
        sales = float(row[2])

        if row[3].strip():
            age = int(float(row[3]))
            total_age += age
        total_sales += sales
        customer_count += 1
average_age = total_age / customer_count
print(f"Total Sales: ${total_sales}")
print(f"Average Customer Age: {average_age}")
```

5 - Advanced Data Analysis with Pandas:


*   Group data by product and calculate total sales per product.

*   Identify and list all unique payment methods used.

```python
import pandas as pd
# Load the CSV file
df = pd.read_csv('assignment_sales_data.csv')

# Group by product and calculate total sales
product_sales = df.groupby('Product')['Sales'].sum()
print("Total Sales by Product:")
print(product_sales)

print("========================================================================")
# Identify and list all unique payment methods used.
unique_payment_methods = df['Payment_Method'].dropna().unique()
print("Unique Payment Methods:")
print(unique_payment_methods)
```

4 - Generate a Summary Report:
    Create a summary report in a text file or Word document that includes:
*   Total sales
*   Average customer age
*   Total sales by product
*   List of unique payment methods    

```python
with open('SedrickASS21_IETL_summary_report.txt', 'w') as report:
    report.write("===== SALES SUMMARY REPORT =====\n\n")
    report.write(f"Total Sales: ${total_sales:.2f}\n")
    report.write(f"Average Customer Age: {average_age:.2f}\n\n")

    report.write("Total Sales by Product:\n")
    for product, sales in product_sales.items():
        report.write(f" - {product}: ${sales:.2f}\n")

    report.write("\nUnique Payment Methods:\n")
    for method in unique_payment_methods:
        report.write(f" - {method}\n")

print("Summary report generated: SedrickASS21_IETL_summary_report.txt")
```
