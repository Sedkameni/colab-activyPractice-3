```python
# Import Pandas library
import pandas as pd
import requests
```

```python
### Task 1: Working with Pandas read_html

# URL of a webpage with finance data
finance_url = "https://finance.yahoo.com/gainers"
head = {"User-agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:140.0) Gecko/20100101 Firefox/140.0"}
response = requests.get(finance_url, headers=head)
```

```python
### 2 - Extract tables from the URL using Pandas read_html() into a DataFrame.
# Read HTML tables from the webpage
finance_tables = pd.read_html(response.text)
```

```python
## 3 -Save only the first table into another DataFrame..
print("First table:")
print(finance_tables[0])
df=finance_tables[0]
```
