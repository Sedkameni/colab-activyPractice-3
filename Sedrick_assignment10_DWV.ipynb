```python
#### 1 - Copy these lines of code into your Python file
import pandas as pd
import matplotlib.pyplot as plt

# Load Titanic dataset
titanic_data = pd.read_csv("Titanic-Dataset.csv")

# Clean the dataset (handling missing values)
titanic_data.dropna(inplace=True)

# Display the first five rows
print(titanic_data.head())
```

```python
### 2.1 - Line Plot: Show the trend of age distribution among passengers.
# Sort the ages to show a trend
sorted_ages = titanic_data['Age'].sort_values().values

# Create a line plot
plt.figure(figsize=(10, 5))
plt.plot(sorted_ages, color='green', linewidth=2)
plt.title('Trend of Age Distribution Among Passengers')
plt.xlabel('Passenger Index (Sorted by Age)')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

```python
# Count passengers per ticket class
class_counts = titanic_data['Pclass'].value_counts().sort_index()

# Create a vertical bar plot
plt.figure(figsize=(8, 6))
plt.bar(class_counts.index, class_counts.values, color='skyblue', edgecolor='black')
# Add labels and title
plt.title('Passenger Count by Ticket Class')
plt.xlabel('Ticket Class')
plt.ylabel('Number of Passengers')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
### 2.3 - Bar Plot (Horizontal): Show the count of passengers in each embarkation port.
# Count passengers per embarkation port
embark_counts = titanic_data['Embarked'].value_counts()

# Create a horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(embark_counts.index, embark_counts.values, color='lightgreen', edgecolor='black')

# Add labels and title
plt.title('Passenger Count by Embarkation Port')
plt.xlabel('Number of Passengers')
plt.ylabel('Embarkation Port')
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
### - 2.4 - Pie Chart: Show the percentage of passengers in each ticket class.
plt.figure(figsize=(8, 8))

# Create a pie chart
wedges, texts, autotexts = plt.pie(class_counts.values, labels=class_counts.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgreen', 'lightcoral'], explode=[0.04, 0.04, 0.04])

# Add legend instead of labels
plt.legend(wedges,
           [f'Class {cls}' for cls in class_counts.index],
           title="Ticket Class",
           loc="upper right")
# Add title
plt.title('Percentage of Passengers by Ticket Class')

# Show the plot
plt.show()
```

```python
### 2.5 - Histogram: Display the distribution of passenger ages.
#Size of the frame
plt.figure(figsize=(10, 6))

# Create a histogram of passenger ages
plt.hist(titanic_data['Age'], bins=20, color='c', edgecolor='k', alpha=0.7)

# Add labels and title
plt.title('Distribution of passenger ages')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')

# Show the plot
plt.show()
```

```python
### 2.6 - Scatter Plot: Explore the relationship between age and fare paid.
plt.figure(figsize=(12, 6))

# Create a scatter plot
plt.scatter(titanic_data['Age'], titanic_data['Fare'], color='m', marker='*')

# Add labels and title
plt.title('Scatter Plot of Age & Fare paid')
plt.xlabel('Age')
plt.ylabel('Fare paid')

# Show the plot
plt.show()
```

```python
### 2.7 - Horizontal Box Plot: Visualize the distribution of passenger ages.

# Create a boxplot plot
plt.figure(figsize=(8, 6))
plt.boxplot(titanic_data['Age'], vert=False)

# Add labels and title
plt.title('Box Plot of the distribution of passenger ages')
plt.xlabel('Passenger ages')

# Show the plot
plt.show()
```

```python
#### 2.8 - Violin Plot: Compare the distribution of ages for passengers who didnt survive vs. those who did.
ages_not_survived = titanic_data[titanic_data['Survived'] == 0]['Age']
ages_survived = titanic_data[titanic_data['Survived'] == 1]['Age']

# Create violin plot
plt.figure(figsize=(8, 6))
plt.violinplot([ages_not_survived,ages_survived], showmeans=True)

# Add labels and title
plt.title('Age Distribution by Survival Status')
plt.xlabel('Survival Status')
plt.ylabel('Age')
plt.xticks([1, 2], ['Did Not Survive', 'Survived'])
plt.grid(axis='y', linestyle='--', alpha=0.6)


plt.show()
```

```python
#### 2.9 - Area Plot: Show the cumulative fare paid over time.

# Calculate cumulative fare
titanic_data['CumulativeFare'] = titanic_data['Fare'].cumsum()

# Create area plot
plt.figure(figsize=(12, 6))
plt.fill_between(titanic_data.index, titanic_data['CumulativeFare'], color='skyblue', alpha=0.5)
plt.plot(titanic_data.index, titanic_data['CumulativeFare'], color='Slateblue', alpha=0.6, linewidth=2)

# Add labels and title
plt.title('Cumulative Fare Paid Over Time')
plt.xlabel('Passenger Index')
plt.ylabel('Cumulative Fare')
plt.grid(linestyle='--', alpha=0.6)

# Show the plot
plt.show()
```

```python
#### 2.10 - Error Bar Plot: Display the mean and standard deviation of passenger ages.
# Calculate mean and standard deviation of passenger ages
mean_age = titanic_data['Age'].mean()
std_age = titanic_data['Age'].std()

# Create error bar plot
plt.figure(figsize=(6, 6))
plt.errorbar(x=['All Passengers'], y=[mean_age], yerr=[std_age], fmt='o', color='blue', ecolor='red', elinewidth=2, capsize=5,  label='Mean Age  Std Dev')

# Add labels, title, and legend
plt.title('Mean Passenger Age with Standard Deviation')
plt.ylabel('Age')
plt.xlabel('Passenger Group')
plt.ylim(0, 80)   # Set y-limit for clarity
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
