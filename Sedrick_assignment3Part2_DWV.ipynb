```python
#### Task 2: Scrape content using BeautifulSoup and lxml (xml) parser

### 1 -  Install Python BeautifulSoup and Requests Library
import requests
from bs4 import BeautifulSoup

# Use this xml code
menu_xml = '''
<?xml version="1.0" encoding="UTF-8"?>
<breakfast_menu>
  <food>
    <name>Belgian Waffles</name>
    <price>5.95</price>
    <description>Two of our famous Belgian Waffles with plenty of real maple syrup</description>
    <calories>650</calories>
  </food>
  <food>
    <name>Strawberry Belgian Waffles</name>
    <price>7.95</price>
    <description>Light Belgian waffles covered with strawberries and whipped cream</description>
    <calories>900</calories>
  </food>
  <food>
    <name>Berry-Berry Belgian Waffles</name>
    <price>8.95</price>
    <description>Light Belgian waffles covered with an assortment of fresh berries and whipped cream</description>
    <calories>900</calories>
  </food>
  <food>
    <name>French Toast</name>
    <price>4.50</price>
    <description>Thick slices made from our homemade sourdough bread</description>
    <calories>600</calories>
  </food>
  <food>
    <name>Homestyle Breakfast</name>
    <price>6.95</price>
    <description>Two eggs, bacon or sausage, toast, and our ever-popular hash browns</description>
    <calories>950</calories>
  </food>
</breakfast_menu>
'''
```

```python
### 2.2 - Scrap content using BeautifulSoup and lxml (xml) parser.
soup = BeautifulSoup(menu_xml, 'lxml')
```

```python
### 2.3 - Find all food information.
foods = soup.find_all('food')
print(foods)
```

```python
##### 2.4 - Load each food item into Python objects.
for food in foods:
    name = food.find('name').text
    price = float(food.find('price').text)
    description = food.find('description').text
    calories = int(food.find('calories').text)
```

```python
#### 2.5 - Display all food items for each food.

for food in foods:
    name = food.find('name').text
    price = float(food.find('price').text)
    description = food.find('description').text
    calories = int(food.find('calories').text)
    print(f"Name: {name}")
    print(f"Price: ${price:.2f}")  # 2f converts to 2 float points
    print(f"Descriptions: {description}")
    print(f"Calories: {calories}")
    print()   # Inputs a next line after each drink
```
