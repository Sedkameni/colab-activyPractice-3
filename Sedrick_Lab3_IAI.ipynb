```python
# Save the dataset to a new CSV file
df.to_csv("patients.csv", index=False)
# Read it back to verify
df_csv = pd.read_csv("patients.csv")
print(df_csv.head())
```

```python
import pickle

# Save the DataFrame
with open("patients.pkl", "wb") as f:
    pickle.dump(df, f)

# Load the DataFrame back from the pickle file
with open("patients.pkl", "rb") as f:
    df_pickle = pickle.load(f)

# Now you can safely print it
print(df_pickle.head())

```

```python
import sqlite3

# Connect to (or create) database
conn = sqlite3.connect("patients.db")
cursor = conn.cursor()
# Save DataFrame to SQLite table
df.to_sql("patients", conn, if_exists="replace", index=False)
```

```python
# Query first 5 rows from database
df_sql = pd.read_sql("SELECT * FROM patients LIMIT 5", conn)
print(df_sql)
# Close connection
conn.close()
!git add Lesson_3_Lab_python_refresher.ipynb
!git commit -m "Commit 3: File I/O, Pickle, and SQLite database operations
```

```python
# import all the similar files, csv's

import zipfile
import os

# Path to your datasets zip
zip_path = "datasets.zip"

# Check if the zip file exists
if os.path.exists(zip_path):
    print("datasets.zip found.")

    # List all CSV files inside the zip (without extracting)
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        csv_files_in_zip = [f for f in zip_ref.namelist() if f.endswith('.csv')]
        print("CSV files inside datasets.zip:")
        for file in csv_files_in_zip:
            print(" -", file)

    # Extract all files
    !unzip -o datasets.zip -d datasets
    print("\n Files extracted successfully to 'datasets' folder.")
# List CSV files after extraction
    extracted_csvs = [f for f in os.listdir('datasets') if f.endswith('.csv')]
    print("\n CSV files available after extraction:")
    for file in extracted_csvs:
        print(" -", file)

else:
    print(" datasets.zip not found in this folder.")



```
