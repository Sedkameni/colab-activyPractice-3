```python
## 3 - Reload file, specify data has no header. Skip first 7 rows and show the next 12 rows after. Save to a new dataframe and print.
new_df=pd.read_csv('employees.csv', header=None, skiprows=7, nrows=12)
print(new_df)
```

```python
#### Part 2 : Load from and read to EXCEL file
#####  1 - Import pandas library.
import pandas as pd

##### 2 - Using ExcelFile, load "file1.xlsx" (available at the bottom of this page) into a dataframe.
file1=pd.ExcelFile("file1.xlsx")
```

```python
##### 3 - Display the sheet names.
file1.sheet_names
```

```python
##### 4 - Display Sheet1.
sheet1=file1.parse(sheet_name='Sheet1')
print(sheet1)
```

```python
### 5 - Display Sheet1 and specify the first column as the index column.
new_sheet1=file1.parse(sheet_name='Sheet1', index_col=0)
print(new_sheet1)
```

```python
#### 6 - Display Sheet1 and skip the first and third columns.
new_sheet12=file1.parse(sheet_name='Sheet1', usecols=[1, 3, 4, 5])
print(new_sheet12)
```

```python
##### 7 -  Using ExcelWriter and to_excel, save to a new excel file named “file_sample” the following:
##############  xlsample2 to a sheet called “Sample1”.
##############   xlsample3 to a sheet called “Sample2”.
sheet2=sheet1.copy()
with pd.ExcelWriter('file_sample.xlsx') as writer:
    new_sheet1.to_excel(writer, sheet_name='Sample1')
    new_sheet12.to_excel(writer, sheet_name='Sample2')
```

```python
##### Part 3 - Load from and read to JSON file
### 1 - Import pandas library.
import pandas as pd

##### 2 - Load "sample.json" (available at the bottom of this page) file data into a dataframe and print.
dfjson = pd.read_json('sample.json')
print(dfjson)
```

```python
#### 3 - Create a dictionary with the details of one customer.
import json

with open('sample.json', 'r') as f:
    datadict = json.load(f)
print(datadict)
```

```python
### 4 - Convert to a dataframe.
df=pd.DataFrame(datadict)
```

```python
#### 5 - Print as a Json string
print(df.to_json())
```

```python
#### 6 - Export as a Json file.
df.to_json('new_sample.json')
```
