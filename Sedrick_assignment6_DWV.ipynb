```python
## - Import this file by specifying the filename and filepath in the pd.read_csv() function.
import pandas as pd
food_sales = pd.read_csv('Food_Sales.csv')
## Display the first 5 rows to view data.
print("The 5 first rows are:", food_sales.head())
```

```python
### - Part A - Data Mapping
category = {"Carrot": "Bars", "Chocolate Chip": "Cookies", "Arrowroot": "Cookies", "Oatmeal Raisin": "Cookies",
            "Bran": "Bars", "Whole Wheat": "Crackers", "Potato Chips": "Snacks", "Pretzels": "Snacks", "Banana": "Bars"}
#### - Map Product column to category and save to a column, Category.
food_sales["Category"] = food_sales["Product"].map(category)

## - Print the first 10 rows.
print(food_sales.head(10))
```

```python
### - Part B - Discretization and Binning
## - Define your number of bins to 3.
bins_numb = 3

## - Bin with equal width the column Unit_Price and save to a new column Price_Bins.
food_sales["Price_Bins"] = pd.cut(food_sales["Unit_Price"],  bins = bins_numb, labels = False)

## - Bin with equal frequency the column Unit_Price and save to a new column Price_Category,
###using labels properties [‘Low’, ‘Medium’, ‘High’].
food_sales["Price_Category"] = pd.qcut(food_sales["Unit_Price"], q = bins_numb, labels = ["Low", "Medium", "High"])

## - Print the first 5 rows.
print(food_sales.head(5))
```

```python
### Part C - Data Type Conversion
### - Check the datatype of your dataframe.
print("\nThe data types are:\n", food_sales.dtypes)

#### - Convert datatype for Price_Category column to a string datatype.
food_sales["Price_Category"] = food_sales["Price_Category"].astype(str)

#### - Check the datatype of your dataframe after datatype conversion.
print("\nThe data type after conversion of the Price_category type to string are:\n", food_sales.dtypes)
```

```python
### Part D - Random Sampling
### ------- Import random -----
import random

#### - Randomly sample 10 rows of your dataframe. Save to a Python object and print.
random_sample = food_sales.sample(n = 10)
print(random_sample)
```

```python
#### Part E - Data Aggregation
## - Group by City column and the sum of Quantity column. Save to a Python object and print.
city_quantity = food_sales.groupby("City")["Quantity"].sum()
print("\nThe Quatity grouped by city are:\n", city_quantity)

## - Group by Product column and compute aggregation functions (mean, min, max, column) of Quantity column.
##Save to a Python object and print.
product_quantity = food_sales.groupby("Product").agg({"Quantity": ["mean", "min", "max"]})
print("\nThe Quatity grouped by product are:\n", product_quantity)

## - Create a new column in dataframe, Total_Price from the multiplication of Unit_price and Quantity columns.
## Using transform, group by Product and calculate the mean Total_Price,
##then add this mean total price to your dataframe as a column called Mean_Price. Display the first 5 rows of dataframe.

## - create Total_Price from the multiplication of Unit_price and Quantity columns
food_sales["Total_Price"] = food_sales["Unit_Price"] * food_sales["Quantity"]

## Group by 'Mean_Price' and calculate the mean Total_Price
food_sales["Mean_Price"] = food_sales.groupby("Product")["Total_Price"].transform("mean")

## - Create a custom function to calculate percentage total using column, Total_Price.
def calc_amount(food_sales):
    total_price = food_sales['Total_Price'].sum()
    food_sales['Percentage_Total'] = (food_sales['Total_Price'] / total_price) * 100
    return food_sales

##Using apply function, apply created function to each group defined by Product column.
result = food_sales.groupby('Product').apply(calc_amount, include_groups=False)
##Print first 5 rows of result.
print("\nThe data set after performing creation of Total_Price column and percentage total is:\n", result.head())
```

```python
### Part F - Data Transformation
#### ------- Import StandardScaler -------
from sklearn.preprocessing import StandardScaler

#### - Scale numeric values (Unit_Price, Quantity, Total_Price, Mean_Price)
scaler = StandardScaler()
food_sales[["Unit_Price_s", "Quantity_s", "Total_Price_s", "Mean_Price_s"]] = scaler.fit_transform(food_sales[["Unit_Price", "Quantity", "Total_Price", "Mean_Price"]])

##### using StandardScaler fit.transform function. Display the first 5 rows of dataframe.
print("\nThe data set after scaling numeric values is:\n", food_sales.head())
```

```python
##### Part F - Data Transformation
#### - Convert categorial variables to dummy using get_dummies of your dataframe.
dummies_food_sales = pd.get_dummies(food_sales) # , columns = ["Category", "Price_Category"]

#### - Display the first 5 rows of dataframe.
print("\nThe data set after converting categorial variables to dummy is:\n", dummies_food_sales.head())

##Save to a Python object and convert Boolean True/False values to 0/1 using astype(int).
dummies_food_sales = dummies_food_sales.astype(int)
##Display the first 5 rows of dataframe.
print("\nThe data set after converting Boolean True/False values to 0/1 is:\n", dummies_food_sales.head())
```
