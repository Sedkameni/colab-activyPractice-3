```python
### Part 1 - Solve the below task. (NOTE: Download bigMart.csv and save it to your file directory.)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

bigmart = pd.read_csv('bigMart.csv')
print(bigmart.head())
print(bigmart.shape)
print(bigmart.info())
print(bigmart.describe())
print(bigmart.isnull().sum())
```

```python
### 1.2 - Setting plots to have a white grid.
sns.set_style('whitegrid')
```

```python
### 2-Visualization Tasks: (NOTE: For each task, customize plots by adding a title, x-label, and y-label)
### 2.1 - Catplot (boxplot): Create a catplot (box) to show the relationship between Outlet_Type and Item_Outlet_Sales.
####      Parameters to explore: Outlet_Location_Type as hue, palette, height.
plt.figure(figsize=(10, 6))
sns.catplot(data=bigmart, x='Outlet_Type', y='Item_Outlet_Sales', kind='box', hue='Outlet_Location_Type', palette='Set3', height=8)
plt.title('Box Plot of Outlet_Type vs Item_Outlet_Sales')
plt.xlabel('Outlet_Type')
plt.ylabel('Item_Outlet_Sales')
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

```python
#### 2.2 - Violinplot(): Create a violionplot to show the relationship between Outlet_Location_Type and Item_Outlet_Sales.
####       Parameters to explore: Item_Fat_Content as hue, True to notch, palette, and width.
plt.figure(figsize=(10, 6))
sns.violinplot(data=bigmart, x='Outlet_Location_Type', y='Item_Outlet_Sales', palette='cool', hue='Item_Fat_Content', split=True, width=0.9) #, notch=True
plt.title('Violin Plot of Outlet_Location_Type vs Item_Outlet_Sales')
plt.xlabel('Outlet_Location_Type')
plt.ylabel('Item_Outlet_Sales')
plt.legend(title='Item_Fat_Content', loc=2)
plt.tight_layout()
plt.show()
```

```python
##### 2.3 - Countplot(): Create a countplot to show the number of occurrences for each Outlet_Type.
#####       Parameters to explore: Outlet_Location_Type as hue, and palette.
plt.figure(figsize=(10, 6))
sns.countplot(data=bigmart, x='Outlet_Type', hue='Outlet_Location_Type', palette='Set2')
plt.title('Count Plot of Outlet_Type')
plt.xlabel('Outlet_Type')
plt.ylabel('Count')
plt.legend(title='Outlet_Location_Type')
plt.show()
```

```python
#### 2.4 - Boxenplot(): Create a boxenplot to show the relationship between Item_Type and Item_Outlet_Sales.
####       Parameters to explore: Item_Fat_Content as hue, full to k_depth, and palette.
plt.figure(figsize=(12, 8))
sns.boxenplot(data=bigmart, x='Item_Type', y='Item_Outlet_Sales', hue='Item_Fat_Content', palette='Set3', k_depth='full')
plt.title('Boxen Plot of Item_Type vs Item_Outlet_Sales')
plt.xlabel('Item_Type')
plt.ylabel('Item_Outlet_Sales')
plt.legend(title='Item_Fat_Content', loc=2)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#### 2.5 - Stripplot(): Create a stripplot to show the relationship between Outlet_Size and Item_Visibility.
#####      Parameters to explore: Item_Fat_Content as hue, True to jitter, True to dodge, palette, and size.
plt.figure(figsize=(10, 6))
sns.stripplot(data=bigmart, x='Outlet_Size', y='Item_Visibility', hue='Item_Fat_Content', jitter=True, dodge=True, palette='Set2', size=8)
plt.title('Strip Plot of Outlet_Size vs Item_Visibility')
plt.xlabel('Outlet_Size')
plt.ylabel('Item_Visibility')
plt.legend(title='Item_Fat_Content', loc=2)
plt.show()
```
