```python
import requests
url = "https://jsonplaceholder.typicode.com/posts/"
# Add a payload dictionary
payload = {"id": [1, 2, 3], "userId":1}
# Get request to API
response = requests.get(url, params=payload)
# Print the response
response_json = response.json()
for i in response_json:
  print(i, "\n")
```

```python
from lxml import html
import requests
# Define the URL
url = 'https://example.com'
# Send an HTTP request to the website and retrieve the HTML content
response = requests.get(url)
# Parse the HTML content using lxml
tree = html.fromstring(response.content)
# Extract titles from the HTML tree using XPath
link_titles = tree.xpath('//a/text()')
# Print the extracted link titles
for title in link_titles:
  print(title)
```

```python
html = """
<head>
<title class="page-title">Hello World!</title>
</head>
<body>
<div id="content">
<h1>Title</h1>
<p>first paragraph</p>
<p>second paragraph</p>
<h2>Subtitle</h2>
<p>first paragraph of subtitle</p>
</div>
</body>
"""
# copy and use this HTML code in previous slide
from bs4 import BeautifulSoup
# Build a soup object from html text
soup = BeautifulSoup(html, 'lxml')
# Display title node
print(soup.head.title)
# Displays the whole HTML node text
print(soup.head.title.text)
# Display the class attributes
print(soup.head.title["class"])
```

```python
food_xml = '''
<?xml version="1.0" encoding="UTF-8"?>
<menu>
<food>
<id>1</id>
<name>Strawberry Waffles</name>
<price>$5.95</price>
<description>Light waffles covered with strawberries and whipped cream</description>
<calories>700</calories>
</food><food>
<id>2</id>
<name>French Toast</name>
<price>$4.50</price>
<description>Thick slices made from our homemade sourdough bread</description>
<calories>600</calories>
</food>
</menu>
'''
from bs4 import BeautifulSoup
# Parse the XML content
soup = BeautifulSoup(food_xml, 'xml')
# Find all food items
foods = soup.find_all('food')
# Loop through each food and extract details
for food in foods:
  id = food.find('id').text
  name = food.find('name').text
  price = food.find('price').text
  description = food.find('description').text.strip() # Use strip() to remove leading/trailing white spaces
  calories = food.find('calories').text
  print(f"Food ID: {id}")
  print(f"Name: {name}")
  print(f"Price: {price}")
  print(f"Description: {description}")
  print(f"Calories: {calories}")
  print("-" * 40) # Print a separator line for better readability
```
