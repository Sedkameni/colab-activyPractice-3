```python
import pandas as pd
### Part 2: Load the file: study_performance.csv into a DataFrame.
data = 'study_performance.csv'
## Read the csv file
df = pd.read_csv(data)


##Part 2: 1 - Display the first 20 rows.
print(df.head(20))
```

```python
### Part 2: 2 - Display the last 3 rows.
print(df.tail(3))
```

```python
### Part2: 3 - Display the statistical summary.
print(df.describe())
```

```python
##Part2: 4 - Display the mean, minimum, and maximum values of ‘reading_score’ column.
print("The mean of the reading score is:", df['reading_score'].mean())
print("The minimum value of the reading score is:", df['reading_score'].min())
print("The maximum value of the reading score is:", df['reading_score'].max())
```

```python
### Part 2: 5 - Display the total non-empty entries and datatypes of the columns.
print(df.info())
```

```python
### Part 2: 6 - Display the value count of ‘race_ethnicity’ and ‘parental_level_of_education’ columns.
print("The value count of the race ethnicity and parental level of education are:")
print(df[['race_ethnicity','parental_level_of_education']].value_counts())
```
