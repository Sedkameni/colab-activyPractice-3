```python
#### 1 - Import libraries. Load and explore the "adult.csv" dataset available at the bottom of this page.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
column_names = [
    "age", "workclass", "fnlwgt", "education", "education_num",
    "marital_status", "occupation", "relationship", "race",
    "sex", "capital_gain", "capital_loss", "hours_per_week",
    "native_country", "income"
]

data = pd.read_csv(url, header=None, names=column_names, na_values=' ?', skipinitialspace=True)

# Display first few rows
print(data.head())

# Describe the dataset
print(data.describe())
```

```python
### 2 - Set plots style to white grid.
# Settting plots to have a white grid
sns.set(style="whitegrid")
```

```python
### 3 - Relplot: Create a scatter plot to visualize the relationship between age and hours worked per week, colored by income.
# Plotting scatterplot with relplot
sns.relplot(data=data, x='age', y='hours_per_week', hue='income', kind='scatter', height=6, aspect=1.5, alpha=0.7)
plt.title('Scatter Plot of Age vs. Hours Worked per Week')
plt.xlabel('Age')
plt.ylabel('Hours per Week')
plt.show()
```

```python
#### 4 - Create a line plot to visualize the average hours worked per week by age, separated by income.
# Calculating average hours worked per week by age and income
avg_hours_per_age_income = data.groupby(['age', 'income'])['hours_per_week'].mean().reset_index()

# Plotting line plot using relplot
sns.relplot(data=avg_hours_per_age_income, x='age', y='hours_per_week', hue='income', kind='line', height=6, aspect=1.5, marker='o')
plt.title('Line Plot of Average Hours Worked per Week by Age')
plt.xlabel('Age')
plt.ylabel('Average Hours per Week')
plt.show()
```

```python
#### 5 - Scatterplot: Create a scatter plot to visualize the relationship between capital gain and capital loss, colored by income.
# Plotting scatterplot using scatterplot()
sns.relplot(data=data, x='capital_gain', y='capital_loss', hue='income', kind='scatter', height=6, aspect=1.5, alpha=0.7)
plt.title('Scatter Plot of Capital Gain vs. Capital Loss')
plt.xlabel('Capital Gain')
plt.ylabel('Capital Loss')
plt.legend(title='Income')
plt.show()
```

```python
### 6 - Lineplot: Create a line plot to show the average hours worked per week by age, separated by sex.
# Calculating average hours worked per week by age and sex
avg_hours_per_age_sex = data.groupby(['age', 'sex'])['hours_per_week'].mean().reset_index()

# Plotting lineplot using lineplot()
sns.lineplot(data=avg_hours_per_age_sex, x='age', y='hours_per_week', hue='sex', marker='o')
plt.title('Line Plot of Average Hours Worked per Week by Age and Sex')
plt.xlabel('Age')
plt.ylabel('Average Hours per Week')
plt.show()
```

```python
### 7 - Displot: Create a histogram to show the distribution of ages in the dataset.
# Plotting histogram using displot()
sns.displot(data=data, x='age', bins=30, kde=False, height=6, aspect=1.5, color='green')
plt.title('Histogram of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
##### 8 - Create a KDE plot to show the distribution of ages in the dataset.
# Plotting kdeplot using displot()
sns.displot(data=data, x='age', kind='kde', height=6, aspect=1.5, fill=True, color='purple')
plt.title('KDE Plot of Ages')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python
#### 9 - Create an ECDF plot to show the empirical cumulative distribution of ages in the dataset.
# Plotting ecdfplot using displot()
sns.displot(data=data, x='age', kind='ecdf', height=6, aspect=1.5, color='red')
plt.title('ECDF Plot of Ages')
plt.xlabel('Age')
plt.ylabel('ECDF')
plt.show()
```

```python
#### 10 - Histplot: Create a histogram to show the distribution of hours worked per week.
# Plotting histogram using hisplot()
sns.histplot(data=data, x='hours_per_week', bins=30, color='blue')
plt.title('Histogram with KDE of Hours Worked per Week')
plt.xlabel('Hours per Week')
plt.ylabel('Frequency')
plt.legend(title='Year') #,labels=years
plt.show()
```

```python
##### 11 - Create a histogram to show the distribution of hours worked per week and setting kde to True.
# Plotting histogram using hisplot(), set kde to True
sns.histplot(data=data, x='hours_per_week', bins=30, kde=True, color='orange')
plt.title('Histogram with KDE of Hours Worked per Week')
plt.xlabel('Hours per Week')
plt.ylabel('Frequency')
plt.show()
```

```python
#### 12 - Kdeplot: Create a KDE plot to show the distribution of hours worked per week
# Plotting kdeplot using kdeplot()
sns.kdeplot(data=data, x='hours_per_week', fill=True, color='cyan') ### shade=True,
plt.title('KDE Plot of Hours Worked per Week')
plt.xlabel('Hours per Week')
plt.ylabel('Density')
plt.show()
```

```python
### 13 - Ecdfplot: Create an ECDF plot to show the empirical cumulative distribution of hours worked per week.
# Plotting ecdfplot using ecdfplot()
sns.ecdfplot(data=data, x='hours_per_week', color='magenta')
plt.title('ECDF Plot of Hours Worked per Week')
plt.xlabel('Hours per Week')
plt.ylabel('ECDF')
plt.show()
```

```python
#### 14 - Rugplot: Create a rug plot to show the distribution of ages in the dataset.
# Plotting rugplot
sns.rugplot(data=data, x='age', height=0.05, color='green')
plt.title('Rug Plot of Ages')
plt.xlabel('Age')
plt.show()
```
