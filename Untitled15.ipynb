# PROJET SOMMATIF : ÉCOSYSTÈMES ET DURABILITÉ
## Sciences et technologie - 7e année

---

## PRÉSENTATION DU PROJET

### Contexte
Votre municipalité souhaite aménager un nouveau parc écologique et elle sollicite l'expertise de jeunes scientifiques pour créer un écosystème équilibré et durable. Votre équipe doit concevoir, planifier et présenter un écosystème viable qui respecte les principes écologiques tout en répondant aux besoins de la communauté.

### Attentes du curriculum
**Attente générale B2** : Démontrer sa compréhension de la structure et de la fonction des écosystèmes, des interactions au sein d'un écosystème et des mesures à prendre pour protéger les écosystèmes et préserver la biodiversité.

**Attente générale A1** : Utiliser la démarche d'enquête scientifique et les habiletés connexes pour mener des expériences portant sur les interactions dans un écosystème.

### Durée totale : 4 semaines (12 périodes de 60 minutes)

---

## ÉTAPES DU PROJET

### ÉTAPE 1 : RECHERCHE ET PLANIFICATION (Semaines 1-2)
**Durée** : 5 périodes

**Tâches à accomplir** :
- Choisir un type d'écosystème parmi : forêt tempérée, prairie, zone humide, étang
- Rechercher les composantes biotiques et abiotiques de cet écosystème
- Identifier les chaînes alimentaires et les réseaux trophiques
- Analyser les facteurs qui affectent l'équilibre de l'écosystème
- Rédiger un plan de conception détaillé

**Productions attendues** :
1. Journal de recherche collaboratif (Google Doc partagé)
2. Schéma annoté de l'écosystème avec légende
3. Plan de conception incluant :
   - Liste des organismes (producteurs, consommateurs, décomposeurs)
   - Diagramme du réseau trophique
   - Description des facteurs abiotiques
   - Prédiction des interactions écologiques

---

### ÉTAPE 2 : CONSTRUCTION DU MODÈLE 3D (Semaine 2)
**Durée** : 3 périodes

**Tâches à accomplir** :
- Construire un modèle tridimensionnel de l'écosystème (40 cm x 30 cm minimum)
- Représenter fidèlement les composantes biotiques et abiotiques
- Intégrer des éléments interactifs ou mobiles
- Créer des étiquettes scientifiques pour chaque composante

**Production attendue** :
- Modèle 3D de l'écosystème avec étiquettes explicatives

**Matériaux suggérés** :
- Matériaux recyclés (carton, bouteilles, tissus)
- Éléments naturels (branches, pierres, sable)
- Pâte à modeler ou argile
- Peinture, marqueurs, colle

---

### ÉTAPE 3 : ENQUÊTE SCIENTIFIQUE - PERTURBATION DE L'ÉCOSYSTÈME (Semaine 3)
**Durée** : 2 périodes

**Tâches à accomplir** :
- Choisir UNE perturbation à analyser : pollution, espèce envahissante, changement climatique, destruction d'habitat
- Formuler une hypothèse sur l'impact de cette perturbation
- Simuler la perturbation sur le modèle
- Observer et documenter les effets en cascade
- Proposer des solutions de restauration

**Production attendue** :
- Rapport de laboratoire scientifique comprenant :
  - Hypothèse
  - Méthodologie
  - Observations (photos avant/après)
  - Analyse des impacts
  - Solutions proposées

---

### ÉTAPE 4 : PRÉSENTATION ORALE ET DÉFENSE (Semaine 4)
**Durée** : 2 périodes (10 minutes par équipe : 7 min présentation + 3 min questions)

**Tâches à accomplir** :
- Préparer une présentation PowerPoint/Google Slides (8-12 diapositives)
- Présenter l'écosystème à la classe
- Expliquer le réseau trophique et les interactions
- Démontrer la perturbation et les solutions
- Répondre aux questions de l'auditoire

**Production attendue** :
- Présentation visuelle (numérique ou affiche)
- Présentation orale structurée

## TRIANGULATION DE L'ÉVALUATION

### 1. OBSERVATION (25% de l'évaluation sommative)

#### Quoi observer ?
- Collaboration et répartition du travail en équipe
- Utilisation de la démarche d'enquête scientifique
- Manipulation sécuritaire du matériel
- Persévérance face aux défis techniques
- Communication en français pendant le travail

#### Quand ?
- Continuellement durant les 4 semaines
- Moments clés : périodes de recherche, construction, expérimentation

#### Outils d'observation :
- **Grille d'observation continue** (voir page suivante)
- **Fiches anecdotiques** pour noter des comportements significatifs
- **Photos/vidéos** des élèves en action (avec permission)

---

### 2. CONVERSATION (25% de l'évaluation sommative)

#### Types de conversations :

**A) Conférences d'équipe** (Semaine 2 et 3)
- Rencontre de 7-10 minutes avec chaque équipe
- Questions à poser :
  - "Expliquez-moi comment l'énergie circule dans votre écosystème."
  - "Que se passerait-il si on enlevait tous les décomposeurs ?"
  - "Pourquoi avez-vous choisi ces organismes spécifiques ?"
  - "Comment votre écosystème maintient-il son équilibre ?"
  - "Quelle est la différence entre chaîne alimentaire et réseau trophique ?"

**B) Questionnement formatif individuel** (Tout au long)
- Questions spontanées pendant la circulation en classe
- Évaluation de la compréhension conceptuelle
- Utilisation du vocabulaire scientifique

**C) Présentation orale finale** (Semaine 4)
- Présentation structurée de 7 minutes
- Période de questions de 3 minutes

#### Outils de conversation :
- **Grille d'évaluation de la conférence** (voir page suivante)
- **Notes prises pendant les échanges**
- **Enregistrement audio** des présentations (optionnel)

---

### 3. PRODUCTION (50% de l'évaluation sommative)

#### Productions évaluées :

**A) Journal de recherche collaboratif** (10%)
- Complétude des recherches
- Organisation de l'information
- Utilisation de sources crédibles
- Vocabulaire scientifique approprié

**B) Plan de conception + Schémas** (10%)
- Clarté et précision du plan
- Qualité des schémas annotés
- Identification correcte des composantes
- Représentation précise du réseau trophique

**C) Modèle 3D de l'écosystème** (15%)
- Exactitude scientifique
- Créativité et esthétique
- Fonctionnalité (éléments mobiles/interactifs)
- Qualité de la construction

**D) Rapport scientifique sur la perturbation** (10%)
- Respect du format de rapport de laboratoire
- Qualité de l'hypothèse et de l'analyse
- Pertinence des solutions proposées
- Liens avec les concepts écologiques

**E) Présentation visuelle (PowerPoint/Affiche)** (5%)
- Clarté et organisation
- Qualité visuelle
- Intégration de schémas/photos
- Absence de fautes de français

- Permettre l'utilisation d'un dictionnaire bilingue

- Offrir des glossaires illustrés du vocabulaire scientifique

## EXEMPLE DE RAPPORT SCIENTIFIQUE ANNOTÉ

### RAPPORT DE LABORATOIRE : PERTURBATION D'UN ÉCOSYSTÈME

**Titre :** Impact d'une espèce envahissante sur l'écosystème de la forêt tempérée

**Introduction** (2-3 phrases)
*Expliquez le contexte de votre expérience et les concepts scientifiques pertinents.*

*Exemple :* "Notre écosystème représente une forêt tempérée d'Ontario. Dans cette enquête, nous avons simulé l'arrivée d'une espèce envahissante, la carpe asiatique, dans l'étang de notre forêt. Les espèces envahissantes sont des organismes introduits dans un nouvel écosystème où ils n'ont pas de prédateurs naturels, ce qui peut perturber l'équilibre écologique."

**Hypothèse** (1-2 phrases)
*Formulez votre prédiction et justifiez-la.*

*Exemple :* "Nous prédisons que l'introduction de carpes asiatiques dans l'étang causera une diminution des poissons indigènes et des plantes aquatiques. Cette prédiction est basée sur le fait que les carpes asiatiques sont des consommateurs voraces qui entrent en compétition avec les espèces locales pour la nourriture."

**Méthodologie** (liste numérotée des étapes)
1. Nous avons construit notre modèle d'écosystème avec 10 organismes différents incluant des truites et des nénuphars dans l'étang.
2. Nous avons introduit 3 carpes asiatiques (représentées par des poissons en plastique) dans l'étang.
3. Nous avons simulé leur consommation en retirant 50% des plantes aquatiques et 2 truites.
4. Nous avons observé et documenté les changements avec des photos.
5. Nous avons tracé le nouveau réseau trophique après la perturbation.

**Observations** (photos, tableaux, graphiques)

*Photo 1 : Écosystème AVANT la perturbation*
*Photo 2 : Écosystème APRÈS la perturbation*

| Organisme | Population AVANT | Population APRÈS | Changement |
|-----------|------------------|------------------|------------|
| Nénuphars | 10 | 5 | -50% |
| Truites | 4 | 2 | -50% |
| Carpes asiatiques | 0 | 3 | +100% |
| Hérons | 2 | 1 | -50% |

**Analyse** (2-3 paragraphes)
*Expliquez ce que vous avez observé en utilisant les concepts scientifiques.*

*Exemple :* "Nos observations confirment notre hypothèse. L'introduction des carpes asiatiques a causé des effets en cascade dans tout l'écosystème. D'abord, la compétition pour la nourriture a réduit la population de truites indigènes. Ensuite, la diminution des plantes aquatiques a affecté l'oxygénation de l'eau et l'habitat des petits poissons. Finalement, la réduction du nombre de truites a impacté les hérons qui s'en nourrissent, forçant ces prédateurs à chercher de la nourriture ailleurs. Cet exemple démontre l'interconnexion des organismes dans un réseau trophique et l'importance de chaque maillon."

**Discussion - Solutions proposées** (1 paragraphe)
*Proposez des solutions réalistes pour restaurer l'équilibre.*

*Exemple :* "Pour restaurer l'équilibre de cet écosystème, nous proposons trois solutions : 1) retirer les carpes asiatiques en utilisant des filets spécialisés, 2) réintroduire des truites reproductrices, et 3) replanter des nénuphars et autres plantes aquatiques. Il faudrait aussi surveiller l'écosystème pendant plusieurs années pour s'assurer que l'équilibre est rétabli."

**Conclusion** (2-3 phrases)
*Résumez vos découvertes et faites un lien avec un enjeu environnemental réel.*

*Exemple :* "Cette enquête nous a appris que les espèces envahissantes peuvent causer des dommages importants aux écosystèmes locaux. Au Canada, plusieurs espèces envahissantes comme la moule zébrée et l'agrile du frêne menacent nos écosystèmes. Il est crucial de prévenir l'introduction de ces espèces et d'agir rapidement quand elles sont détectées."

**Bibliographie**
- Espace pour la vie. (2024). *Les espèces envahissantes au Québec*. espacepourlavie.ca
- Ministère des Richesses naturelles. (2023). *Guide sur la carpe asiatique*. ontario.ca
