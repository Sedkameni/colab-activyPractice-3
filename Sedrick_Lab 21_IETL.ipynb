Extract Specific Information from the CSV File

    Extract the Total Sales
    Extract the average Customer Age

```python
import csv
total_sales = 0
total_age = 0
customer_count = 0
with open('sales_data.csv', mode='r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)  # Skip the header row
    for row in csv_reader:
        sales = float(row[2])
        age = int(row[3])
        total_sales += sales
        total_age += age
        customer_count += 1
average_age = total_age / customer_count
print(f"Total Sales: ${total_sales}")
print(f"Average Customer Age: {average_age}")
```

4 - Perform Data Cleaning Tasks

    Handle missing values and correct formatting issues

```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('sales_data.csv')
# Fill missing values
df.fillna({'Sales': 0, 'Customer_Age': df['Customer_Age'].mean()}, inplace=True)
# Correct formatting issues (example: standardize customer location names)
df['Customer_Location'] = df['Customer_Location'].str.title()
print(df)
```

5 - Use Pandas for Advanced Data Analysis

    Perform more advanced analysis using pandas, such as grouping and aggregating data5

```python
import pandas as pd
# Load the CSV file
df = pd.read_csv('sales_data.csv')

# Group by product and calculate total sales
product_sales = df.groupby('Product')['Sales'].sum()
print("Total Sales by Product:")
print(product_sales)
# Calculate average sales by customer location
location_sales = df.groupby('Customer_Location')['Sales'].mean()
print("Average Sales by Customer Location:")
print(location_sales)
# Filter data for sales above a certain amount
high_sales = df[df['Sales'] > 1000]
print("Sales Above $1000:")
print(high_sales)
```
