```python
#### Task 3: 	Scrape content using BeautifulSoup and html.parser.

#### 1 -   Install Python BeautifulSoup and Requests Library.
import requests

from bs4 import BeautifulSoup

# HTML content of the adventure
adventure_html = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>The Adventure of the Bear, Rabbit, and Tortoise</title>
</head>
<body>
    <h1 class="title">The Adventure of the Bear, Rabbit, and Tortoise</h1>

    <p>Once upon a time, in a peaceful forest, there lived a bear, a rabbit, and a tortoise.</p>

    <div class="characters">
        <p class="character" id="bear">Bear</p>
        <p class="character" id="rabbit">Rabbit</p>
        <p class="character" id="tortoise">Tortoise</p>
    </div>

    <p>These three friends spent their days exploring the woods, playing games, and enjoying each other's company.</p>

    <div class="story-arc">
        <h2>Chapter 1: The Mysterious Forest</h2>
        <p>One sunny morning, the bear, rabbit, and tortoise stumbled upon a mysterious forest...</p>
    </div>

    <div class="story-arc">
        <h2>Chapter 2: The Hidden Treasure</h2>
        <p>Curiosity piqued, they ventured deeper into the forest in search of the rumored hidden treasure...</p>
    </div>

    <div class="story-arc">
        <h2>Chapter 3: The Great Discovery</h2>
        <p>After overcoming various obstacles, they finally uncovered the long-lost treasure...</p>
    </div>
</body>
</html>
"""
```

```python
### 3.2 - Scrap content using BeautifulSoup and html.parser.
soup = BeautifulSoup(adventure_html, 'html.parser')
```

```python
### 3.3 - Display the head tag..
head_tag = soup.head
print("Head tag:")
print(head_tag)
```

```python
### 3.4 - Display the title tag.
print(soup.title)
```

```python
### 3.5 - Display the text of the title tag
print(soup.title.string)
```

```python
### 3.6 - Display the p class in the body tag.
print(soup.body.p)
```

```python
### 3.7 -  Display the div class in the body tag.
print(soup.body.div)
```

```python
### 3.8 - Display h1 tag.
print(soup.h1)
```

```python
##### 3.9 - Filter based on a string ‘title’ and print the text of the string.
title_tag = soup.find('title')
print("Title of the adventure:", title_tag.text)
```

```python
#### 3.10 - Filter based on a regular expression ‘character’ and print each result.
# Filter attribute based on a regular expression
characters = soup.find(class_=lambda x: x and 'character' in x)
print("\nCharacters in the adventure:")  #\n inserts a new line
for character in characters:
    print(character.text)
```

```python
#### 3.11 - Filter based on a list [“story-arc”] and print each result.
# Filter by attribute based on a list
story_arcs = soup.find_all(class_=["story-arc"])
print("\nAdventure Arcs:")
for arc in story_arcs:
    print(arc.h2.text)
```

```python
#### 3.12 - Filter based on a function and print the result.
# Filter by attribute based on a function
def is_title_tag(tag):
    return tag.name == "h1" and "title" in tag.get('class', [])

title_tag = soup.find(is_title_tag)
print("\nTitle of the adventure using a function:", title_tag.text)
```

```python
#### 3.13 - Filter based on the value True and print each result.
# Filter by attribute with value True
div_with_id = soup.find(id=True).text
print("\nFirst div with an 'id' attribute:", div_with_id)
```
