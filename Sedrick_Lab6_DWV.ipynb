```python
## 2- Install Python Pandas, NumPy, and SciPy Library.
import pandas as pd
```

```python
## 3 - Import "Sales.csv"
sales = pd.read_csv('Sales.csv')

print(sales.head())
```

```python
## 4 - Data Mapping: Create a category dictionary and map to Item column in the dataframe.
category = {"Apple": "Fruits", "Orange": "Fruits", "Banana": "Fruits",
     "Cheese": "Dairy", "Milk": "Dairy", "Bread": "Bakery",
     "Chicken": "Meat", "Beef": "Meat", "Pork": "Meat", "Fish": "Fish"}

# Mapping Item to category
sales["category"] = sales["item"].map(category)
```

```python
## 5 - Replace “Fish” in Category column with “Seafood”.
sales["category"] = sales["category"].replace('Fish', 'Seafood')

#print(duplicate)
print(sales)
```

```python
## 6 - Discretization and Binning: Define the number of bins.
num_bins = 5
```

```python
## 7- Using equal width binning on the Price column and create a new column, Price_Bins.
# Binning with equal width
sales['Price_Bins'] = pd.cut(sales['price'], bins=num_bins, labels=False)
```

```python
## 8 - Using equal frequency binning on the Price column and create a new column, Price_Category. Give labels to the bins.
sales['Price_Category'] = pd.cut(sales['price'], bins=num_bins, labels=['Cheap', 'Fairly Cheap', 'Moderate', 'Fairly Expensive', 'Expensive'])
```

```python
## 9 - Display first 10 rows to see effect.
print(sales.head(10))
```

```python
## 10 - Data Type Conversion: Check datatype for Sales dataframe.
print(sales.dtypes)
```

```python
## 11 - Convert datatype for Sales_Id and Price_Category to a string datatype.
sales['sales_id'] = sales['sales_id'].astype(str)
sales['Price_Category'] = sales['Price_Category'].astype(str)
```

```python
## 12 - Check datatype for Sales dataframe after conversion.
print("The new data types are: ")
print(sales.dtypes)
```

```python
## 13 - Random Sampling: Import random and randomly sample 4 rows in the Sales dataframe. Print result.
import random
random_sample = sales.sample(n=4)
print(random_sample)
```

```python
## 14 - Data Aggregation: Using groupby function, group by Item column and calculate the sum of the Quantity column.
#--------- Data Aggregation--------------
# Group by 'Item' and calculate the sum of 'Quantity'
item_quantity = sales.groupby('item')['quantity'].sum()
print(item_quantity)
```

```python
## 15 - Using groupby function, group by Item column and compute aggregration functions
#(mean, std, min, max, count) on the Quantity column.
# Group by 'Item' and compute multiple aggregation functions
agg_result = sales.groupby('item').agg({'quantity': ['mean', 'std', 'min', 'max', 'count']})
print("\nAggregation Result:")
print(agg_result)
```

```python
## 16 - Calculate Sales_Amount by multiplying Price and Quantity.
## Using transform,  group by Item column and calculate the mean of the Sales_Amount column.
# Create Sales_Amount column from Price * Quantity
sales['Sales_Amount'] = sales['price'] * sales['quantity']

# Group by 'Item' and calculate the mean Sales_Amount
mean_amount= sales.groupby('item')['Sales_Amount'].transform('mean')
```

```python
## 17 - Add Mean_Sales to dataframe and display first 5 rows of result.
# Add the mean amount to the DataFrame
sale['Mean_Sales'] = mean_amount
print(sales.head())
```

```python
## 18 - Calculate Sales_Amount percentage total by custom function.
## Using apply, group by Item column by applying the result of the custom function.
# Define a custom function to calculate the percentage total
def calc_amount(sales):
    total_amount = sales['Sales_Amount'].sum()
    sales['Percentage_Total'] = (sales['Sales_Amount'] / total_amount) * 100
    return sales

# Apply the custom function to each group defined by 'Item'
result = sales.groupby('item').apply(calc_amount, include_groups=False)
print(result.head())
```

```python
## 19 - Data Transformation: Import library and using StandardScaler function, scale columns Price, Quantity, and Sales_Amount.
##      Display the first 5 rows of the resulting dataframe.
#---------Data Transformation---------
from sklearn.preprocessing import StandardScaler

# Scale numeric variables using StandardScaler
scaler = StandardScaler()
sales[['Price', 'Quantity', 'Sales_Amount']] = scaler.fit_transform(sales[['price', 'quantity', 'Sales_Amount']])

print(sales.head())
```

```python
## 20 - Using get_dummies, convert categorical variables into dummies and the Boolean resulting datatype to int.
##      Display the first 5 rows of the resulting dataframe.

# Convert categorical variable into dummy/indicator variables
dummy_sales = pd.get_dummies(sales)

# Convert boolean True/False to integer 0s and 1s
dummy_sales = dummy_sales.astype(int)

print(dummy_sales.head())
```
