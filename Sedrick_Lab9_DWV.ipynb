```python
### 1 - Load and explore the dataset
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
red_wine = pd.read_csv(r'winequality-red.csv', sep=';')
white_wine = pd.read_csv(r'winequality-white.csv', sep=';')

# Display the first few rows and summary statistics
print(red_wine.head())
print(white_wine.head())

```

```python
### 2 - ARGS and Keyword ARGS: Create a scatter plot of alcohol vs. quality using positional arguments for red_wine dataset.
plt.scatter(red_wine['alcohol'], red_wine['quality'])
plt.xlabel('Alcohol')
plt.ylabel('Quality')
plt.title('Red Wine: Alcohol vs. Quality')
plt.show()
```

```python
### 3 - Create a scatter plot of alcohol vs. quality using keyword arguments for red_white dataset.

# Scatter plot using keyword arguments
plt.scatter(x=white_wine['alcohol'], y=white_wine['quality'])
plt.xlabel('Alcohol')
plt.ylabel('Quality')
plt.title('Alcohol vs. Quality (White Wine)')
plt.show()
```

```python
#### 4 - Figures and Subplots: Create a figure with two subplots, one for red wine and one for white wine.
# plt.figure(figsize=(12, 6))
# plt.subplot(1, 2, 1)
# plt.scatter(red_wine['alcohol'], red_wine['quality'])
# plt.xlabel('Alcohol')
# plt.ylabel('Quality')
# plt.title('Red Wine: Alcohol vs. Quality')
# plt.subplot(1, 2, 2)
# plt.scatter(white_wine['alcohol'], white_wine['quality'])
# plt.xlabel('Alcohol')
# plt.ylabel('Quality')
# plt.title('White Wine: Alcohol vs. Quality')

# Creating figure and subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
```

```python
### 5 - Use a scatter plot to visualize the relationship between alcohol and quality for ax1 subplots using red_wine dataset and define color to red.
######  Add title and labels for x and y.
#fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
# Creating a scatter plot for red_wine dataset
ax1.scatter(red_wine['alcohol'], red_wine['quality'], color='red', alpha=0.5)

# Adding title and labels for x and y axis.
ax1.set_title('Red Wine')
ax1.set_xlabel('Alcohol')
ax1.set_ylabel('Quality')


```

```python
#### 6 - Use a scatter plot to visualize the relationship between alcohol and quality for ax2 subplots using white_wine dataset and define color to red.
######   Add title and labels for x and y.
# Creating a scatter plot for white_wine dataset
ax2.scatter(white_wine['alcohol'], white_wine['quality'], color='white', alpha=0.5, edgecolor='black')

# Adding title and labels for x and y axis.
ax2.set_title('White Wine')
ax2.set_xlabel('Alcohol')
ax2.set_ylabel('Quality')
```

```python
### 7 - Using plt.suptitle() define a super heading for the subplots. Display the plot.
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
ax1.scatter(red_wine['alcohol'], red_wine['quality'], color='red', alpha=0.5)

# Adding title and labels for x and y axis.
ax1.set_title('Red Wine')
ax1.set_xlabel('Alcohol')
ax1.set_ylabel('Quality')

ax2.scatter(white_wine['alcohol'], white_wine['quality'], color='white', alpha=0.5, edgecolor='black')

# Adding title and labels for x and y axis.
ax2.set_title('White Wine')
ax2.set_xlabel('Alcohol')
ax2.set_ylabel('Quality')
plt.suptitle('Alcohol vs. Quality for Red and White Wines')
plt.show()
```

```python
### 8 - Colors, Markers, and Line Styles: Group by alcohol and the mean value of quality for each dataset.
# Group by alcohol and mean value of quality
average_quality_red = red_wine.groupby('alcohol')['quality'].mean()
average_quality_white = white_wine.groupby('alcohol')['quality'].mean()
```

```python
#### 9 - Create a line plot showing the average quality score for different levels of alcohol content.
### Customize the plot using different colors, markers, and line styles for each plot. Display the plot.
plt.figure(figsize=(10, 6))
# Creating a line plot with custom colors, markers and line styles for each dataset.
plt.plot(average_quality_red.index, average_quality_red.values, color='red', marker='o', linestyle='-', label='Red Wine')
plt.plot(average_quality_white.index, average_quality_white.values, color='blue', marker='x', linestyle='--', label='White Wine')

plt.show()
```

```python
### 10 - Ticks, Labels, and Legends
plt.figure(figsize=(10, 6))
plt.plot(average_quality_red.index, average_quality_red.values, color='red', marker='o', linestyle='-', label='Red Wine')
plt.plot(average_quality_white.index, average_quality_white.values, color='blue', marker='x', linestyle='--', label='White Wine')
##### 11 - Add labels for x and y with a font size of 12 each.
########## Add a title with a font size of 14 and a font weight as bold. and legend.
plt.xlabel('Alcohol Content', fontsize=12)
plt.ylabel('Average Quality Score', fontsize=12)
plt.title('Average Quality Score by Alcohol Content', fontsize=14, fontweight='bold')

##### 12 - Add ticks for the x and y axis with a font size of 10 each. Add legend. Display the plot.
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.legend()
plt.show()
```

```python
#### 13 - Legends and Grids
plt.figure(figsize=(15, 9))
plt.plot(average_quality_red.index, average_quality_red.values, color='red', marker='o', linestyle='-', label='Red Wine')
plt.plot(average_quality_white.index, average_quality_white.values, color='blue', marker='x', linestyle='--', label='White Wine')

plt.xlabel('Alcohol Content', fontsize=12)
plt.ylabel('Average Quality Score', fontsize=12)
plt.title('Average Quality Score by Alcohol Content', fontsize=14, fontweight='bold')

### 14 - Add a grid with color, line style, and linewidth features. Add a legend with loc and font size. Display the plot.
plt.grid(color='gray', linestyle='--', linewidth=0.5)

plt.legend(loc=4, fontsize=10)
plt.show()
```

```python
### 15 - Annotations and Drawings on a Subplot
plt.figure(figsize=(15, 9))
plt.plot(average_quality_red.index, average_quality_red.values, color='red', marker='o', linestyle='-', label='Red Wine')
plt.plot(average_quality_white.index, average_quality_white.values, color='blue', marker='x', linestyle='--', label='White Wine')

##### 16 - Find the minimum quality value for alcohol and score of the red_wine dataset.
min_quality_alcohol = average_quality_red.idxmin()
min_quality_score = average_quality_red.min()

##### 17 - Annotate the plot by marking the point with the lowest quality score.
#######    Draw an arrow pointing to this point and add a text annotation.

plt.annotate('Lowest Quality', xy=(min_quality_alcohol, min_quality_score), xytext=(min_quality_alcohol+1, min_quality_score+0.5),
             arrowprops=dict(facecolor='brown', shrink=0.05),
             fontsize=12, color='green')

plt.xlabel('Alcohol Content', fontsize=12)
plt.ylabel('Average Quality Score', fontsize=12)
plt.title('Average Quality Score by Alcohol Content', fontsize=14, fontweight='bold')
plt.suptitle('Wine Quality Analysis', fontsize=16, fontweight='bold', color='purple')

plt.grid(color='gray', linestyle='--', linewidth=0.5)

plt.legend(loc=4,fontsize=10)

#### 18 - Saving Plots as a File: Save the plot as a PNG and a PDF file into your desired file path. Display the plot.
# Save plot as PNG and PDF
plt.savefig('wine_quality_plot.png')
plt.savefig('wine_quality_plot.pdf')
plt.show()
```
