```python
##### 1 - Loading and cleaning dataset

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

## Load the dataset “planets” from Seaborn databases.
planets = sns.load_dataset("planets")

###  Clean data of missing values
planets = planets.dropna()

### Explore the dataset.

# Display first few rows
print(planets.head())

# Describe the dataset
print(planets.describe())

## Set grid style to white (PS: This is optional)
sns.set(style="whitegrid")
```

```python
### 2 - Visualization Tasks:
#### 2.1 - Relplot: Visualize the relationship between two variables, “distance” and “orbital_period”,
####       and set these parameters: “scatter” to kind, “number” to hue, and choose a desired palette.
sns.relplot(data=planets, x='distance', y='orbital_period', hue='number', kind='scatter', palette = None, height=6, aspect=1.5, alpha=0.7)
plt.title('Scatter Plot of distance vs orbital_period')
plt.xlabel('distance')
plt.ylabel('orbital_period')
plt.show()
```

```python
##### 2.2 - Relplot: Visualize the relationship between two variables, “year” and “orbital_period”, and set these parameters:
#####       “line” to kind, “number” to hue, and choose a desired palette, marker, and line style.
sns.relplot(data=planets, x='year', y='orbital_period', hue='number', kind='line', palette = 'magma', marker = '*', ls =':', height=6, aspect=1.5, alpha=0.7)
plt.title('Line Plot of year vs orbital_period')
plt.xlabel('year')
plt.ylabel('orbital_period')
plt.show()
```

```python
### 2.3 - Scatterplot: Visualize the relationship between two variables, “orbital_period” and “mass”, and set these parameters:
####     “number” to hue, 0.7 to alpha, and choose a desired palette.
sns.scatterplot(data=planets, x='orbital_period', y='mass', hue='number', alpha=0.7, palette = 'viridis')
plt.title('Scatter Plot of orbital_period vs mass')
plt.xlabel('orbital_period')
plt.ylabel('mass')
plt.legend(title='number')
plt.show()
```

```python
#### 2.4 - Lineplot: Visualize the relationship between two variables, “year” and “distance”, and set these parameters: color to blue, choose a desired marker, and line style.
sns.lineplot(data=planets, x='year', y='distance', color='blue', marker='o', ls='--')
plt.title('Line Plot of year vs distance')
plt.xlabel('year')
plt.ylabel('distance')
plt.show()
```

```python
#### 2.5 - Displot: Explore the distribution of a single variable “mass”. Set the color parameter to green.
sns.displot(data=planets, x='mass', bins=30, kde=False, height=6, aspect=1.5, color='green')
plt.title('Distribution Plot of mass')
plt.xlabel('mass')
plt.ylabel('Frequency')
plt.show()

```

```python
### 2.6 - Displot: Explore the distribution of a single variable “year”. Set parameters: “number” to hue, kde as True, and choose a desired palette.
sns.displot(data=planets, x='year', hue='number', kde=True, palette='magma', height=6, aspect=1.5, alpha=0.7)
plt.title('Distribution Plot of year')
plt.xlabel('year')
plt.ylabel('Frequency')
plt.show()
```

```python
#### 2.7 - Histplot: Visualize the distribution of a single variable “orbital_period”. Set parameters: “number” to hue, 20 to bins, and choose a desired palette.
sns.histplot(data=planets, x='orbital_period', hue='number', bins=20, palette='viridis')
plt.title('Histogram Plot of orbital_period')
plt.xlabel('orbital_period')
plt.ylabel('Frequency')
plt.show()
```

```python
### 2.8 - KDEplot: Display the probability density function of a variable, “orbital_period”. Set parameters: “number” to hue, fill as True, and choose a desired palette.
sns.kdeplot(data=planets, x='orbital_period', hue='number', fill=True, palette='magma')
plt.title('Kernel Density Plot of orbital_period')
plt.xlabel('orbital_period')
plt.ylabel('Density')
plt.show()
```

```python
#### 2.9 - EDCFplot: Display the probability density function of a variable, “orbital_period”. Set parameters: “number” to hue.
sns.ecdfplot(data=planets, x='orbital_period', hue='number')
plt.title('Empirical Cumulative Distribution Function Plot of orbital_period')
plt.xlabel('orbital_period')
plt.ylabel('ECDF')
plt.show()
```

```python
#### 2.10 - Rugplot: Display the probability density function of a variable, “orbital_period”. Set parameters: “number” to hue, and 0.5 to height.
sns.rugplot(data=planets, x='orbital_period', hue='number', height=0.5)
plt.title('Rug Plot of orbital_period')
plt.xlabel('orbital_period')
plt.show()
```
