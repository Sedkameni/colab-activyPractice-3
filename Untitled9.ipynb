```python
### 2 - Load and explore the dataset
import pandas as pd

# Load the dataset
penguins = pd.read_csv('penguins.csv')

# Display first few rows
print(penguins.head())

# Describe the dataset
print(penguins.describe())
```

```python
#### - Data Reshaping in Pandas: Create a pivot table showing the average body mass for each species on each island.
# Pivoting
pivot_table = penguins.pivot_table(values='body_mass_g', index='species', columns='island', aggfunc='mean')
print(pivot_table)
```

```python
### - Melt the pivot table back to a long format.
# Melting
melted_table = pivot_table.reset_index().melt(id_vars='species', value_vars=pivot_table.columns)
print(melted_table)
```

```python
### - Stack the dataframe and convert it to a multi-index series.
# Stacking
stacked_table = pivot_table.stack()
print(stacked_table)
```

```python
### - Unstack the multi-index series back to a dataframe.
# Unstacking
unstacked_table = stacked_table.unstack()
print(unstacked_table)
```

```python
### - Categorical Extension Type in Pandas: Convert the species column to a categorical data type using the Categorical constructor.
# Convert species to categorical using Categorical constructor
penguins['species'] = pd.Categorical(penguins['species'])
print(penguins['species'].dtype)
```
