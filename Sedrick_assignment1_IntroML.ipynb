 **Load the dataset using Pandas.**

```python
import pandas as pd

# Load dataset from CSV file
df = pd.read_csv("fruit_classification_dataset.csv")

# Display the first few rows of the dataset
print(df.head())
```

**Preprocess the Data**

    1 - Normalizing the weight values.
    2 - Encoding categorical variables such as color
    3 - Removing irrelevant features
    


    
    
    



```python
from sklearn.preprocessing import LabelEncoder

# Handle missing values
print("The Missing values are:\n", df.isnull().sum())
df = df.dropna()

# Normalizing the weight values.
df["Weight (grams)"] = df["Weight (grams)"] / 1000
# Encode categorical variables
le = LabelEncoder()
df["Color"] = le.fit_transform(df["Color"])
```

**Split Data into Training and Testing Sets**

   1 -  Select relevant features (color, Weight).

  2 - Define the target variable (Type of fruit).
  
  3 - Split the data into training and testing sets.


```python
from sklearn.model_selection import train_test_split

# Define features and target variable
X = df[["Color", "Weight (grams)"]]
y = df["Type"]

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

**Train the Decision Tree Model**

1 - Initialize the Decision Tree classifier.

2 - Train the model using the training set.


```python
from sklearn.tree import DecisionTreeClassifier

# Train Decision Tree Classifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

**Make Predictions**

1 - Use the trained model to make predictions on the test set.
    
2 - Display the predictions.


```python
# Predict on test data
y_pred = model.predict(X_test)
print("Predictions:", y_pred)
```

**Evaluate Model Performance**

1 - Calculate accuracy using the test set.

2 - Display a classification report.


```python
from sklearn.metrics import accuracy_score, classification_report

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Display classification report
print(classification_report(y_test, y_pred))
```
