```python
##### Part 2:  Enter code below after importing pandas:
import pandas as pd
people={
            "firstName": ["Joe","James","Emily"],
            "lastName": ["Jackson","Smith","Jones"],
            "gender": ["male","male","female"],
            "age": [28,32,24],
            "maritalStatus": ["married","single","single"]
}
```

```python
##### Convert to a DataFrame and specify the column names.
df=pd.DataFrame(people,columns=["firstName","lastName","gender","age","maritalStatus"])
print(df)
```

```python
##### Print as a JSON string.
print(df.to_json())
```

```python
##### Export as a JSON file named ‘people.json’.
df.to_json('people.json')
```
