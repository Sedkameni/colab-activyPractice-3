```python
### 1 - Load and explore the dataset:
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('countries-aggregated.csv')

print("The five first rows of the data set are:\n", df.head())

print("\nThe statistics of the data set is:\n", df.describe())
```

```python
### 2 - Line Plot: Create a line plot showing the global confirmed cases over time.
# Filter data for a specific country, e.g., "World"
world_data = df.groupby('Date').sum().reset_index()

print(world_data.head())

plt.figure(figsize=(12, 6))
plt.plot(world_data['Date'], world_data['Confirmed'], label='Confirmed Cases')
plt.title('Global Confirmed COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
### - Bar Plot (Vertical and Horizontal): Create a vertical bar plot showing the confirmed cases for the first 12 days.

# Creating a Vertical bar plot
plt.figure(figsize=(10, 6))
plt.bar(world_data['Date'][:12], world_data['Confirmed'][:12], color='b')
plt.title('Confirmed COVID-19 Cases for the First 12 Days')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.xticks(rotation=45)
plt.show()
```

```python
##### 4 - Create a horizontal bar plot for the same data.
plt.figure(figsize=(10, 6))
plt.barh(world_data['Date'][:12], world_data['Confirmed'][:12], color='g')
plt.title('Confirmed COVID-19 Cases for the First 12 Days')
plt.xlabel('Confirmed Cases')
plt.ylabel('Date')
plt.show()
```

```python
### 5 - Multiseries Bar Chart: Create a multiseries bar chart showing confirmed and recovered cases for the first 12 days.
x = world_data['Date'][:12]
confirmed = world_data['Confirmed'][:12]
recovered = world_data['Recovered'][:12]

plt.figure(figsize=(12, 6))
bar_width = 0.4
index = range(len(x))

# Creating a multiseries bar chart
plt.bar(index, confirmed, bar_width, label='Confirmed Cases')
plt.bar([i + bar_width for i in index], recovered, bar_width, label='Recovered Cases')

plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.title('Confirmed and Recovered COVID-19 Cases for the First 12 Days')
plt.xticks([i + bar_width / 2 for i in index], x, rotation=45)
plt.legend()
plt.show()
```

```python
#### 6 - Stacked Bar Chart: Create a stacked bar chart showing confirmed and recovered cases for the first 12 days.
plt.figure(figsize=(10, 6))
plt.bar(world_data['Date'][:12], confirmed, label='Confirmed Cases')
plt.bar(world_data['Date'][:12], recovered, bottom=confirmed, label='Recovered Cases')

plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.title('Stacked Bar Chart of COVID-19 Cases')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
### 7 - Pie Chart: Create a pie chart showing the proportion of confirmed cases for the first 6 days.
labels = world_data['Date'][:6]
sizes = world_data['Confirmed'][:6]

# Creating a pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Confirmed COVID-19 Cases for the First 6 Days')
plt.show()
```

```python
#### 8 - Histogram: Create a histogram showing the distribution of confirmed cases.
plt.figure(figsize=(10, 6))
plt.hist(world_data['Confirmed'], bins=20, color='c', edgecolor='k', alpha=0.7)
plt.title('Distribution of Confirmed COVID-19 Cases')
plt.xlabel('Confirmed Cases')
plt.ylabel('Frequency')
plt.show()
```

```python
#### 9 - Scatter Plot: Create a scatter plot showing the relationship between date and confirmed cases.
plt.figure(figsize=(12, 6))
plt.scatter(world_data['Date'], world_data['Confirmed'], color='m', marker='*')
plt.title('Scatter Plot of Date vs Confirmed Cases')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.xticks(rotation=45)
plt.show()
```

```python
### 10 - BoxPlot: Create a horizontal box plot showing the distribution of confirmed cases.
plt.figure(figsize=(8, 6))
plt.boxplot(world_data['Confirmed'], vert=False)
plt.title('Box Plot of Confirmed COVID-19 Cases')
plt.xlabel('Confirmed Cases')
plt.show()
```

```python
#### 11 - Violin Plot: Create a violin plot showing the distribution of confirmed cases.
plt.figure(figsize=(8, 6))
plt.violinplot(world_data['Confirmed'], showmeans=True, showmedians=True)
plt.title('Violin Plot of Confirmed COVID-19 Cases')
plt.xlabel('Confirmed Cases')
plt.show()
```

```python

plt.figure(figsize=(12, 6))
plt.fill_between(world_data['Date'], world_data['Confirmed'], color='y', alpha=0.5)
```

```python
#### 12 - Area Plot: Create an area plot showing the confirmed cases over time.
plt.figure(figsize=(12, 6))
plt.fill_between(world_data['Date'], world_data['Confirmed'], color='skyblue', alpha=0.4)
plt.plot(world_data['Date'], world_data['Confirmed'], color='Slateblue', alpha=0.6, linewidth=2)
plt.title('Area Plot of Confirmed COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.xticks(rotation=45)
plt.show()
```

```python
#### 13 - Error Bar Plot: Create an error bar plot showing the confirmed cases with standard deviation for the first 12 days.
# Calculate actual standard deviation for the first 12 days
std_dev = world_data['Confirmed'][:12].std()

# Create the error bar plot
confirmed = world_data['Confirmed'][:12]
dates = world_data['Date'][:12]

plt.figure(figsize=(12, 6))
plt.errorbar(dates, confirmed, yerr=std_dev, fmt='o', ecolor='r', capsize=5, linestyle='--', markersize=5)
plt.title('Error Bar Plot of Confirmed COVID-19 Cases for the First 12 Days')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.xticks(rotation=45)
plt.show()
```
