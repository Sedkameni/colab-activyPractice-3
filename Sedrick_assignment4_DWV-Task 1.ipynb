```python
### Task 1: Work with SQlAlchemy.
## import libraries
#import pandas as pd
import sqlalchemy as db
```

```python
### 1.2 - Create a database â€“ Store.db
#This is done by just specifying the database name in the after filepath e.g. filepath\Store.db


#Create an engine
engine = db.create_engine('sqlite:///Store.db')

#Create a connection
conn = engine.connect()

# Create an instance of MetaData
metadata = db.MetaData()
```

```python
#### 1. 3 - Create tables:
Product = db.Table('Product', metadata,
db.Column('Product_Id', db.Integer(),primary_key=True),
db.Column('Product_Name', db.String(255), nullable=False),
db.Column('Price', db.Integer()),
db.Column('Aisle_Id', db.String(2), nullable=False))

order = db.Table('Order', metadata,
db.Column('Order_Id', db.Integer(),primary_key=True),
db.Column('Customer_Name', db.String(255), nullable=False),
db.Column('Amount', db.Integer()),
db.Column('Product_Id', db.Integer(), nullable=False))

aisle = db.Table('Aisle', metadata,
db.Column('Aisle_Id', db.String(2),primary_key=True),
db.Column('Aisle_Name', db.String(255), nullable=False),
db.Column('Product_Id', db.Integer(), nullable=False))
```

```python
#### 1.4 - Insert data into these tables:

# Create changes in database
metadata.create_all(engine)
   # Insert into Product
#query = db.insert(Product)
conn.execute(Product.insert(), [
    {'Product_Id': 1, 'Product_Name': 'Apple', 'Price': 1, 'Aisle_Id': 'A1'},
    {'Product_Id': 2, 'Product_Name': 'Bread', 'Price': 2, 'Aisle_Id': 'A2'},
    {'Product_Id': 3, 'Product_Name': 'Milk', 'Price': 3, 'Aisle_Id': 'A3'},
    {'Product_Id': 4, 'Product_Name': 'Eggs', 'Price': 4, 'Aisle_Id': 'A1'},
    {'Product_Id': 5, 'Product_Name': 'Juice', 'Price': 5, 'Aisle_Id': 'A2'},
    {'Product_Id': 6, 'Product_Name': 'Cereal', 'Price': 6, 'Aisle_Id': 'A4'}])

# Insert into Order
conn.execute(order.insert(), [
    {'Order_Id': 101, 'Customer_Name': 'Alice', 'Amount': 2, 'Product_Id': '1'},
    {'Order_Id': 102, 'Customer_Name': 'Bob', 'Amount': 1, 'Product_Id': '2'},
    {'Order_Id': 103, 'Customer_Name': 'Charlie', 'Amount': 3, 'Product_Id': '3'},
    {'Order_Id': 104, 'Customer_Name': 'David', 'Amount': 1, 'Product_Id': '1'},
    {'Order_Id': 105, 'Customer_Name': 'Ella', 'Amount': 2, 'Product_Id': '5'},
    {'Order_Id': 106, 'Customer_Name': 'Frank', 'Amount': 1, 'Product_Id': '4'},
    {'Order_Id': 107, 'Customer_Name': 'Grace', 'Amount': 2, 'Product_Id': '6'},
    {'Order_Id': 108, 'Customer_Name': 'Helen', 'Amount': 1, 'Product_Id': '3'}
])

# Insert into Aisle
conn.execute(aisle.insert(), [
    {'Aisle_Id': 'A1', 'Aisle_Name': 'Fruits', 'Product_Id': '1'},
    {'Aisle_Id': 'A2', 'Aisle_Name': 'Bakery', 'Product_Id': '2'},
    {'Aisle_Id': 'A3', 'Aisle_Name': 'Dairy', 'Product_Id': '3'},
    {'Aisle_Id': 'A4', 'Aisle_Name': 'Cereal & Grains', 'Product_Id': '6'}
])


#results = conn.execute(query,values)
```
