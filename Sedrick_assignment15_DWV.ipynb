```python
## Task 1 - Loading dataset:
## Import libraries.
import pandas as pd
import plotly.express as px
## Load database  using pd.read_csv()
url = 'https://raw.githubusercontent.com/plotly/datasets/master/2011_us_ag_exports.csv'
df  = pd.read_csv(url)
## Explore the dataset.
print(df.head())
print(df.info())
print(df.describe)
```

```python
##### Task 2 - Visualization Tasks:
#### 2.1 -  Scatter plot: Create a scatter plot to visualize the relationship between total exports and beef.
#####       Set these parameters: “state” to color. Add title and axis labels.
scatter_fig = px.scatter(df, x='total exports', y='beef', color='state', size='dairy',
                         title='Scatter Plot of total exports vs beef', labels={'total exports': 'total exports', 'beef': 'beef'}) ## hover_name='country', animation_frame='year'
scatter_fig.show()
```

```python
### 2.2 - Bubble plot: Create a bubble chart to show total exports vs poultry with bubble sizes representing total fruits.
####      Set these parameters: “state” to color, 60 to size_max. Add title and axis labels.
bubble_fig = px.scatter(df, x='total exports', y='poultry', color='state', size='total fruits',
                        size_max=60, title='Bubble Plot of total exports vs poultry', labels={'total exports': 'total exports', 'poultry': 'poultry'})
bubble_fig.show()

```

```python
#### 2.3 - Line plot: Create an area plot for total veggies over the states. Add title and axis labels.
line_fig = px.line(df, x='state', y='total veggies', markers=True, title='Line Plot of total veggies over the states', labels={'total veggies': 'total veggies'})
line_fig.show()
```

```python
### 2.4 - Area plot: Create an area plot for total fruits over the states. Add title and axis labels.
area_fig = px.area(df, x='state', y='total fruits', title='Area Plot of total fruits over the states', labels={'total fruits': 'total fruits', 'state': 'state'})

area_fig.show()
```

```python
### 2.5 - Bar plot: Create an area plot for total exports over the states. Add title and axis labels.
bar_fig = px.bar(df, x='state', y='total exports', title='Bar Plot of total exports over the states', labels={'total exports': 'total exports', 'state': 'state'})
bar_fig.show()
```

```python
#### 2.6 - Histogram: Create a histogram for total exports. Add title and axis labels.
hist_fig = px.histogram(df, x='total exports', title='Histogram of total exports', labels={'total exports': 'total exports'})
hist_fig.show()
```

```python
### 2.7 - Tree map: Create a treemap plot to show the hierarchical relationship of state and total exports. Add title.
treemap_fig = px.treemap(df, path=['state'], values='total exports', title='Treemap of state and total exports')
treemap_fig.show()
```

```python
#### 2.8 - Funnel plot: Create a funnel plot for total exports by state. Set parameters: 1000 each to width and height. Add title and axis labels.
funnel_fig = px.funnel(df, x='state', y='total exports', title='total exports by state', labels={'state': 'states', 'total exports':'Total experts'}, width=1000, height = 1000)
funnel_fig.show()
```

```python
#### 2.9 - Heatmap: Create a heatmap for the correlation of all numerical features/ columns. Add title.
df_num = df.select_dtypes(include=['number'])
heatmap_fig = px.imshow(df_num.corr(), title='Correlation Heatmap of Numeric Columns')
heatmap_fig.show()
```

```python
#### 2.10 - Choropleth map: Create a choropleth map to visualize total exports by state.
####        Set parameters: ‘USA-states’ to locationmode, ’state’ to hover_name, ‘usa’ to scope, and ‘Blues’ to color_continuous_scale. Add title.
#################  Hint: locations=’code’ and color=’total exports’
choropleth_fig = px.choropleth(df, locations='code', color='total exports', hover_name='state', scope='usa',
                               title='GDP per Capita by Country', locationmode='USA-states', color_continuous_scale='Blues')
choropleth_fig.show()
```
