```python
#### Task 1 -  Loading dataset

### Import libraries.
import pandas as pd
import plotly.graph_objects as go
#    Load database url = ‘https://raw.githubusercontent.com/plotly/datasets/master/2011_us_ag_exports.csv’ using pd.read_csv()
agexports = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2011_us_ag_exports.csv')
### Explore the dataset.
### Explore the dataset.
print(agexports.head())      # Display the first 5 rows of the dataset to get a glimpse of the data
print(agexports.info())      # Show summary info including data types and non-null counts
print(agexports.describe())  # Generate descriptive statistics like mean, std, min, max for numeric columns
```

```python
### Task 2 - Visualization Tasks (NOTE: For each task, customize plots by adding a title, x-label, and y-label):
### 2.1 - Bubble plot: Create a bubble chart showing the total exports by state, with the size of bubbles representing beef.
             #Hint: The size of the bubbles should be (beef/10)
bubble_fig = go.Figure(go.Scatter(x=agexports['state'], y=agexports['total exports'], mode='markers',
                                  marker=dict(size=agexports['beef']/10, color='red'),
                                  text=agexports['category']))

# Update the layout with title and axis labels
bubble_fig.update_layout(title='total exports with Beef Bubble', xaxis_title='state', yaxis_title='total exports')

# Display the bubble plot
bubble_fig.show()
```

```python
#### 2.2 - Area plot: Create an area plot showing the cumulative total exports over different states.
area_fig = go.Figure(data=go.Scatter(x=agexports['state'],
                                     y=agexports['total exports'],
                                     fill='tozeroy'))

# Update the layout with title and axis labels
area_fig.update_layout(title='Area Plot of total exports over different states',
                       xaxis_title='States', yaxis_title='Total Exports')

# Display the area plot
area_fig.show()
```

```python
### 2.3 - Stacked bar plot: Create a stacked bar plot showing the combined beef and pork exports for each state.
# Create a stacked bar plot for beef and pork exports by state
fig = go.Figure()

# Add beef exports as the first layer of the stack
fig.add_trace(go.Bar(x=agexports['state'], y=agexports['beef'], name='Beef Exports', marker_color='#d62728' ))

# Add pork exports as the second layer of the stack
fig.add_trace(go.Bar(x=agexports['state'], y=agexports['pork'], name='Pork Exports', marker_color='#ff7f0e'))

# Update layout for better visualization
fig.update_layout(
    title={'text': 'Beef and Pork Exports by State (2011)', 'x': 0.5},
    xaxis_title='State',
    yaxis_title='Export Value',
    barmode='stack',  # This creates the stacked effect
    xaxis_tickangle=-45,  # Rotate state names for better readability
    legend=dict(y=0.99, x=0.01)
)

# Show the plot
fig.show()
```

```python
### 2.4 - Table: Create a table showing the total exports for each state.
table_fig = go.Figure(data=[go.Table(header=dict(values=['State', 'Total Exports']), cells=dict(values=[agexports['state'], agexports['total exports']]))])

# Update the layout with a title
table_fig.update_layout(title='Table of total exports for each state')

# Display the table
table_fig.show()
```

```python
#### 2.5 - Candlestick: Create a candlestick plot showing the variation in exports over different states.

candlestick_fig = go.Figure(data=[go.Candlestick(x=agexports['state'],
                                                 open=agexports['total exports'],
                                                 high=agexports['total exports'] + 10000,
                                                 low=agexports['total exports'] - 10000,
                                                 close=agexports['total exports'])])

# Update layout with title and axis labels
candlestick_fig.update_layout(title='Candlestick Plot of total exports over different states', xaxis_title='States', yaxis_title='Total Exports')

# Display the candlestick plot
candlestick_fig.show()
```
