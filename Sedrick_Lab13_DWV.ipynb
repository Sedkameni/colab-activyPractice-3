```python
### Task 1:
### 1 - Import libraries. Load and explore the "mpg.csv" seaborn dataset available at the bottom of this page.
###     You can also use sns.load_dataset("mpg") to load the dataset.

import seaborn as sns
import pandas as pd

# Load the dataset
from matplotlib import pyplot as plt

# Load the dataset
car_data = sns.load_dataset("mpg")

# Display first few rows
print(car_data.head())

# Describe the dataset
print(car_data.describe())

#Settting plots to have a whit grid
sns.set(style="whitegrid")
```

```python
#### 1.2 - Displot: Create a displot with y variable to visualize the distribution of the number of passengers over the years.
# Plotting with y variable for a heatmap-like output
sns.displot(data=car_data, x='horsepower', y='mpg', height=6, aspect=1.5)

plt.title(' Heatmap plot of  Horsepower vs. Miles per Gallon')
plt.xlabel('Horsepower')
plt.ylabel('Miles per Gallon (mpg)')
plt.show()
```

```python
#### 1.3 - Create a displot with y variable and kde to visualize the distribution of the number of passengers over the years.
# Plotting a kdeplot
sns.displot(data=car_data, x='horsepower', y='mpg', kind='kde', height=6, aspect=1.5)

plt.title('KDE Plot of Horsepower vs. Miles per Gallon')
plt.xlabel('Horsepower')
plt.ylabel('Miles per Gallon (mpg)')
plt.show()
```

```python
### 1.4 - Jointplot: Create a joint plot to visualize the relationship between the year and the number of passengers.
# Plotting a jointplot
sns.jointplot(data=car_data, x='weight', y='mpg', kind='scatter', height=8, marginal_kws=dict(bins=30, fill=True))

plt.suptitle('Joint Plot of Weight vs. Miles per Gallon', y=1.02)
plt.xlabel('Weight')
plt.ylabel('Miles per Gallon (mpg)')
plt.show()
```

```python
#### 1.5 - Pairplot: Create a pair plot to visualize the relationships between year, month, and number of passengers.
# Plotting a pairplot
sns.pairplot(data=car_data, vars=['mpg', 'horsepower', 'weight', 'displacement'], hue='origin', palette='coolwarm')

plt.suptitle('Pair Plot of MPG, Horsepower, Weight, and Displacement', y=1.02)
plt.show()
```

```python
### Task 2:
#### 2.1 Import libraries. Load and explore the "flights.csv" seaborn dataset available at the bottom of this page.
#####     You can also use sns.load_dataset("flights") to load the dataset.
# Load the dataset
flights = sns.load_dataset("flights")

# Display first few rows
print(flights.head())

# Describe the dataset
print(flights.describe())
```

```python
#### 2.2 - Regplot: Create a regression plot to visualize the relationship between the year and the number of passengers.
# Plotting a regplot
sns.regplot(data=flights, x='year', y='passengers', scatter_kws={'s': 10}, line_kws={'color': 'red'})

plt.title('Regression Plot of Year vs. Number of Passengers')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
### 2.3 - Lmplot: Create a linear model plot to visualize the relationship between the year and the number of passengers, separated by month.
# Plotting a lmplot
sns.lmplot(data=flights, x='year', y='passengers', hue='month', height=6, aspect=1.5, palette='tab20', markers='o', scatter_kws={'s': 40, 'alpha': 0.6})

plt.title('Linear Model Plot of Year vs. Number of Passengers by Month')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
### 2.4 - Residplot: Create a residual plot to visualize the residuals of the regression of the year on the number of passengers.
# Plotting a residplot
sns.residplot(data=flights, x='year', y='passengers', scatter_kws={'s': 40, 'alpha': 0.6})

plt.title('Residual Plot of Year vs. Number of Passengers')
plt.xlabel('Year')
plt.ylabel('Residuals')
plt.show()

# Create a pivot table for the heatmap and clustermap
flights_pivot = flights.pivot_table(values='passengers', index='month', columns='year', aggfunc='sum', observed=True)
```

```python
### 2.5 - Heatmap: Create a heatmap to visualize the number of passengers per month for each year.
# Plotting a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu", linewidths=0.5, linecolor='gray')

plt.title('Heatmap of Monthly Flight Passengers')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```

```python
### 2.6 - Clustermap: Create a clustermap to visualize the clustering of the number of passengers per month for each year.
# Plotting a clustermap
sns.clustermap(data=flights_pivot, cmap='coolwarm', linewidths=0.5, standard_scale=1, figsize=(12, 8))

plt.title('Clustermap of Number of Passengers per Month and Year')
plt.show()
```

```python
### Task 3:
### 3.1 - Import libraries. Load and explore the "insurance.csv" dataset available at the bottom of this page.
# Load the dataset
url = "https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/insurance.csv"
insurance = pd.read_csv(url)

# Display first few rows
print(insurance.head())

# Describe the dataset
print(insurance.describe())
```

```python
#### 3.2  - Add a grid style
sns.set(style="darkgrid")
```

```python
### 3.3 - Test calling only FacetGrid and the dataset
# Calling FacetGrid only
sns.FacetGrid(insurance)
```

```python
### 3.4 - FacetGrid: Using FacetGrid.map(), create a grid of scatter plots to visualize the relationship between BMI and charges, separated by smoker status. Customize the plot.
# Create a FacetGrid
g = sns.FacetGrid(insurance, col="smoker", height=5, aspect=1.2)
g.map(sns.scatterplot, "bmi", "charges", alpha=0.7)
g.add_legend()

# Set titles and labels
g.set_titles(col_template="Smoker: {col_name}")
g.set_axis_labels("BMI", "Charges")
plt.suptitle('Scatter Plot of BMI vs. Charges by Smoker Status', y=1.05)
plt.show()
```

```python
#### 3.5 - Using FacetGrid.map_dataframe(), create a grid of bar plots to visualize the average charges for each region, separated by sex. Customize the plot.
# Create a FacetGrid
g = sns.FacetGrid(insurance, col="sex", height=5, aspect=1.2)
g.map_dataframe(sns.barplot, x="region", y="charges", palette="viridis")
g.add_legend()

# Set titles and labels
g.set_titles(col_template="Sex: {col_name}")
g.set_axis_labels("Region", "Average Charges")
plt.suptitle('Bar Plot of Average Charges by Region and Sex', y=1.05)
plt.show()
```

```python
### 3.6 - Using FacetGrid with multiple variables, create a grid of histograms to visualize the distribution of charges,
######    separated by region and smoker status. Customize the plot and save the plot as an image.
# Create a FacetGrid
g = sns.FacetGrid(insurance, row="region", col="smoker", height=4, aspect=1.5)
g.map(sns.histplot, "charges", kde=True, color="skyblue")
g.add_legend()

# Set titles and labels
g.set_titles(row_template="{row_name}", col_template="Smoker: {col_name}")
g.set_axis_labels("Charges", "Frequency")
plt.suptitle('Histogram of Charges by Region and Smoker Status', y=1.02)

# Save the figure
g.savefig("FacetGrid.png")
plt.show()
```

```python
#### 3.7 - JointGrid: Using JointGrid.plot(), c reate a joint plot of age vs. charges,
#####      with a scatter plot in the center, histograms along the axes, and customize the color of the scatter points.
# Create JointGrid instance
grid = sns.JointGrid(data=insurance, x='age', y='charges', height=6)

# Plot joint plot
grid.plot(sns.scatterplot, sns.histplot, color='green')
plt.show()
```

```python
#### 3.8 - Using JointGrid.plot() and JointGrid_plot_marginals(), create a joint plot of age vs. charges, with separate scatter plots for smokers and non-smokers,
####       and customize the marker shapes for each group. Add a reference line.
# Create JointGrid instance
grid = sns.JointGrid(data=insurance, x='age', y='charges', height=6)

# Plot joint plot with different markers for smoker and non-smoker
grid.plot_joint(sns.scatterplot, hue=insurance['smoker'], markers=['o', 's'], palette='Set1')

# Plot marginal histograms with different colors for smoker and non-smoker
grid.plot_marginals(sns.histplot, hue=insurance['smoker'], palette='Set1')
grid.refline(x=40, y=35000)

plt.show()
```

```python
#### 3.9 - Using JointGrid.plot() and JointGrid_plot_marginals(), create a joint plot of age vs. charges,
####       with a regression line and kde as True plotted, and customize the line style. Save the plot as an image.

# Create JointGrid instance
grid = sns.JointGrid(data=insurance, x='age', y='charges', height=6)

# Plot joint plot with regression line
grid.plot_joint(sns.regplot, line_kws={'linestyle': '--', 'color': 'red'})

# Plot marginal histograms
grid.plot_marginals(sns.histplot, kde=True)

# Save the plot
grid.savefig('JointGrid.png')

plt.show()
```

```python
#### 3.10 - PairGrid: Using PairGrid.map(), create a pair plot of all numerical variables in the dataset, using scatter plots
# Create PairGrid instance
grid = sns.PairGrid(insurance)

# Plot pair plot with scatter plots
grid.map(sns.scatterplot)

plt.show()
```

```python
##### 3.11 - Using PairGrid.map() and PairGrid.map_offdiag(), create a pair plot of age, charges, and smoker status, using scatter plots in the off-diagonal
#####         and histograms along the diagonal, with different marker shapes for smokers and non-smokers.
# Create PairGrid instance
grid = sns.PairGrid(insurance, vars=['age', 'charges', 'smoker'])

# Plot pair plot with histograms in diagonals
grid.map_diag(sns.histplot)

# Map scatter plot with different marker shapes for smokers and non-smokers
grid.map_offdiag(sns.scatterplot, hue=insurance['smoker'], markers=['o', 's'])

# Show the plot
plt.show()
```

```python
##### 3.12 - Using, PairGrid.map_upper() and PairGrid.map_lower(), create a pair plot of age, charges, and bmi,
#####        with separate regression lines for smokers and non-smokers, and customize the line styles. Save the plot as an image.
# Create PairGrid instance
grid = sns.PairGrid(insurance, vars=['age', 'charges', 'bmi'], height=3)

# Plot pair plot with separate regression lines for smokers and non-smokers
grid.map_upper(sns.regplot, color='purple',  line_kws={'linestyle': '--', 'color': 'green'})
grid.map_diag(sns.histplot)
grid.map_lower(sns.regplot, color='purple',  line_kws={'linestyle': '--', 'color': 'red'})

# Save the plot
grid.savefig('PairGrid.png')

# Show the plot
plt.show()
```
