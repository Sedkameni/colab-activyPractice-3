```python
### Task 1
### 1.1 - Import libraries. Load and explore the "gapminderDataFiveYear.csv" plotly express dataset. You can also use px.data.gapminder() to load this dataset.
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
gapminder = pd.read_csv('gapminderDataFiveYear.csv')
print(gapminder.head())
print(gapminder.info())
print("\nThe Missing values are:\n", gapminder.isnull().sum)
print(gapminder.describe())
```

```python
### 2 - Scatter plot: Create a scatter plot showing life expectancy (lifeExp) vs GDP per capita (gdpPercap) for different countries.
# Scatter Plot
scatter_fig = go.Figure(data=go.Scatter(x=gapminder['gdpPercap'],
                                        y=gapminder['lifeExp'], mode='markers'))
scatter_fig.update_layout(title='Scatter Plot of Life Expectancy vs GDP per Capita', xaxis_title='GDP per Capita', yaxis_title='Life Expectancy')
scatter_fig.show()
```

```python
### 1.3 - Bubble chart: Create a bubble chart showing year vs GDP per capita, with bubble size representing population.
# Bubble chart
bubble_fig = go.Figure(go.Scatter(x=gapminder['year'], y=gapminder['gdpPercap'], mode='markers',
                                  marker=dict(size=gapminder['pop']/1e6, color='red'),
                                  text=gapminder['country']))
bubble_fig.update_layout(title='GDP per Capita with Population Bubble', xaxis_title='Year', yaxis_title='GDP per Capita')
bubble_fig.show()
```

```python
### 1.4 - Line plot: Create a line plot showing life expectancy over time for a specific country.
# Line Plot
line_fig = go.Figure(data=go.Scatter(x=gapminder[gapminder['country'] == 'United States']['year'],
                                     y=gapminder[gapminder['country'] == 'United States']['lifeExp'],
                                     mode='lines'))
line_fig.update_layout(title='Line Plot of Life Expectancy Over Time in the United States',
                       xaxis_title='Year', yaxis_title='Life Expectancy')
line_fig.show()
```

```python
### 1.5 - Text plot: Create a scatter plot with text annotations showing GDP per capita for each country.
# Text Plot
text_fig = go.Figure(data=go.Scatter(x=gapminder['gdpPercap'], y=gapminder['lifeExp'], mode='text', text=gapminder['country']))
text_fig.update_layout(title='Text Plot of GDP per Capita by Country', xaxis_title='GDP per Capita', yaxis_title='Life Expectancy')
text_fig.show()
```

```python
### 1.6 - Area Plot: Create an area plot showing the cumulative life expectancy over time for a specific country.
# Area Plot
area_fig = go.Figure(data=go.Scatter(x=gapminder[gapminder['country'] == 'United States']['year'],
                                     y=gapminder[gapminder['country'] == 'United States']['lifeExp'],
                                     fill='tozeroy'))
area_fig.update_layout(title='Area Plot of Life Expectancy Over Time in the United States',
                       xaxis_title='Year', yaxis_title='Life Expectancy')
area_fig.show()
```

```python
### 1.7 - Bar plot: Create a simple bar plot to visualize the population by continent in the year 2007.
# Filter data for a specific year
gdf = gapminder[gapminder.year == 2007]

# Simple bar plot
fig = go.Figure(go.Bar(x=gdf['continent'], y=gdf['pop'],text=gdf['country'],
     marker=dict(color='blue', opacity=0.7),
))
fig.update_layout(
    title="Population by Continent in 2007",
    xaxis_title="Continent", yaxis_title="Population"
)
fig.show()
```

```python
### 1.8 - Create a grouped bar plot to compare the population by continent for the years 2002 and 2007.
# Filter data for two specific years
df_2002 = gapminder[gapminder.year == 2002]
df_2007 = gapminder[gapminder.year == 2007]

# Grouped bar plot
fig = go.Figure(data=[
    go.Bar(name='2002', x=df_2002['continent'], y=df_2002['pop'], marker=dict(color='green', opacity=0.7)),
    go.Bar(name='2007', x=df_2007['continent'], y=df_2007['pop'], marker=dict(color='blue', opacity=0.7))
])

fig.update_layout(barmode='group',
    title="Population by Continent in 2002 and 2007",
    xaxis_title="Continent", yaxis_title="Population",
)
fig.show()
```

```python
### 1.9 - Create a stacked bar plot to show the population by continent, broken down by country for the year 2007.
# Summarize population by continent and country
continent_pop = gdf.groupby(['continent', 'country'])['pop'].sum().unstack()
print(continent_pop)

# Stacked bar plot
fig = go.Figure()

for country in continent_pop.columns:
    fig.add_trace(go.Bar(x=continent_pop.index, y=continent_pop[country], name=country))

fig.update_layout(
    barmode='stack',
    title="Population by Continent and Country in 2007",
    xaxis_title="Continent", yaxis_title="Population",
)
fig.show()
```

```python
### 1.10 - Pie chart: Create a pie chart showing the distribution of the population among continents.
# Pie Chart
continent_pop = gapminder.groupby('continent')['pop'].sum().reset_index()
pie_fig = go.Figure(data=go.Pie(labels=continent_pop['continent'], values=continent_pop['pop']))
pie_fig.update_layout(title='Pie Chart of Population Distribution by Continent')
pie_fig.show()
```

```python
### 1.11 - Heatmap: Create a heatmap showing the correlation between life expectancy, GDP per capita, and population.
# Heatmap
corr_matrix = gapminder[['lifeExp', 'gdpPercap', 'pop']].corr()
heatmap_fig = go.Figure(data=go.Heatmap(z=corr_matrix.values, x=corr_matrix.columns, y=corr_matrix.columns, colorscale='Viridis'))
heatmap_fig.update_layout(title='Heatmap of Correlation Between Life Expectancy, GDP per Capita, and Population')
heatmap_fig.show()
```

```python
### 1.12 - Table: Create a table showing the GDP per capita for the top 10 countries.
# Table
table_fig = go.Figure(data=[go.Table(header=dict(values=['Country', 'GDP per Capita']),
                                     cells=dict(values=[gapminder['country'], gapminder['gdpPercap']]))])#top10_gdp
table_fig.update_layout(title='Table of GDP per Capita for Top 10 Countries')
table_fig.show()
```

```python
### 1.13 - Violin plot: Create a violin plot showing the distribution of GDP per capita.
# Violin Plot
violin_fig = go.Figure(data=go.Violin(x=gapminder['continent'], y=gapminder['gdpPercap'], box_visible=True, line_color='blue'))
violin_fig.update_layout(title='Violin Plot of GDP per Capita by Continent', xaxis_title='Continent', yaxis_title='GDP per Capita')
violin_fig.show()
```

```python
### 1.14 - Histogram: Create a histogram showing the frequency distribution of GDP per capita.
# Histogram
histogram_fig = go.Figure(data=go.Histogram(x=gapminder['gdpPercap']))
histogram_fig.update_layout(title='Histogram of GDP per Capita', xaxis_title='GDP per Capita', yaxis_title='Count')
histogram_fig.show()
```

```python
### 1.15 - Candlestick: Create a candlestick plot showing the variation in GDP per capita over time for a specific country.
# Task 10: Candlestick
candlestick_fig = go.Figure(data=[go.Candlestick(x=gapminder[gapminder['country'] == 'United States']['year'],
                                                 open=gapminder[gapminder['country'] == 'United States']['gdpPercap'],
                                                 high=gapminder[gapminder['country'] == 'United States']['gdpPercap'] + 1000,
                                                 low=gapminder[gapminder['country'] == 'United States']['gdpPercap'] - 1000,
                                                 close=gapminder[gapminder['country'] == 'United States']['gdpPercap'])])
candlestick_fig.update_layout(title='Candlestick Plot of GDP per Capita Over Time in the United States', xaxis_title='Year', yaxis_title='GDP per Capita')
candlestick_fig.show()
```

```python
### 2.3 - Sankey plot: Create a Sankey plot to visualize the flow between different categories.
# Define the Sankey plot data
labels = ['A', 'B', 'C', 'D', 'E', 'F']
source = [0, 0, 1, 1, 2, 2]
target = [2, 3, 4, 5, 4, 5]
value = [8, 4, 2, 8, 4, 2]

# Create the Sankey plot
fig = go.Figure(go.Sankey(node=dict(pad=15, thickness=20,line=dict(color="black", width=0.5), label=labels),
    link=dict(source=source, target=target, value=value)))

fig.update_layout(title_text="Sankey Diagram Example", font_size=10)
fig.show()
```

```python
### 2.4 - Sunburst plot: Create a Sunburst plot to visualize hierarchical data.
# Define the Sunburst plot data
labels = ["Parent", "Child 1", "Child 2", "Grandchild 1", "Grandchild 2"]
parents = ["", "Parent", "Parent", "Child 1", "Child 1"]
values = [10, 20, 30, 40, 50]

# Create the Sunburst plot
fig = go.Figure(go.Sunburst(labels=labels, parents=parents,
    values=values
))
fig.update_layout(title_text="Sunburst Diagram Example")
fig.show()
```

```python
### 2.5 - Scatter mapbox: Create a Scatter Mapbox plot to visualize geographical data points.
# Define the Scatter mapbox plot data
lon = [-77.0369, -122.4194, -74.0059, -118.2437]
lat = [38.9072, 37.7749, 40.7128, 34.0522]
text = ["Washington DC", "San Francisco", "New York", "Los Angeles"]

# Create the Scatter mapbox plot
fig = go.Figure(go.Scattermapbox(
    lat=lat, lon=lon, mode='markers',
    marker=go.scattermapbox.Marker(size=14,
        color='rgb(255, 0, 0)', opacity=0.7),
    text=text
))

fig.update_layout(mapbox_style="open-street-map",
    mapbox=dict(zoom=3,
        center=dict(lat=39.8283, lon=-98.5795)
    ),
    title="Scatter Mapbox Example"
)
fig.show()
```
