## - Print our dataset

```python
## 3 - Data Manipulation with Pandas:

 #Import pandas library
import pandas as pd

# Create a sample dataset using a dictionary
data = {
    "Name": ["Alice", "Bob", "Charlie", "Diana"],
    "Age": [25, 30, 22, 28],
    "City": ["New York", "Los Angeles", "Chicago", "Houston"],
    "Hobby": ["Reading", "Cycling", "Cooking", "Painting"]
}

# Convert the dictionary into a Pandas DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(" The created dataset is: ", df)
```

This **dataset** contains basic information about four individuals, including their Name, Age, City, and a Hobby they enjoy. It can be used to practice data *manipulation*, *filtering*, and *analysis* with **Pandas**.

#Data visualization

```python
#  4 - Data Visualization with Matplotlib/Seaborn:

# Import libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style
sns.set(style="whitegrid")

# Create a bar chart for Age of each person
plt.figure(figsize=(8,5))
sns.barplot(x="Name", y="Age", data=df, palette="viridis")

# Add title and labels
plt.title("Age of Each Person")
plt.xlabel("Name")
plt.ylabel("Age")

# Show the plot
plt.show()
```

The **bar chart** displays the **ages of four individuals** in our dataset. Each bar represents a person, making it easy to compare their **ages visually**. This type of visualization is useful for quickly understanding differences in numerical values across categories.
