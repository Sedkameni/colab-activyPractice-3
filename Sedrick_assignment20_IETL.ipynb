3- Write code to parse data from the provided XML file.

```python
import xml.etree.ElementTree as ET   # Import ElementTree for parsing XML data

# Parse the XML file and get the root element
tree = ET.parse('assdata.xml')   # Load and parse the XML file
root = tree.getroot()            # Get the root of the XML tree

# Print header row with fixed-width formatting for alignment
print(f"{'Name':<16} {'Age':<5} {'City':<12}")

# Loop through each 'employee' element in the XML
for employee in root.findall('employee'):
    name = employee.find('name').text   # Extract the 'name' element text
    age = employee.find('age').text     # Extract the 'age' element text
    city = employee.find('city').text   # Extract the 'city' element text

    # Print extracted values in aligned table format
    print(f"{name:<16} {age:<5} {city:<12}")
```

*  Extract the names and ages of employees younger than 45.
*   Display the extracted information in a readable format.





*  CSV


```python
import csv   # Import the built-in csv library for parsing CSV files

# Open the CSV file in read mode
with open('assdata.csv', mode='r') as file:
    csv_reader = csv.reader(file)   # Create a CSV reader object

    header = next(csv_reader)   # Skip the header row

    print("Employees younger than 45:")   # Title for the output

    # Loop through each row in the CSV file
    for row in csv_reader:
        # Convert the age column (row[1]) to an integer and check if it is less than 45
        if int(row[1]) < 45:
            # Print the name and age of employees younger than 45
            print(f"Name: {row[0]}, Age: {row[1]}")
```


*   JSON


```python
import json   # Import the built-in json library for parsing JSON files

# Open the JSON file in read mode
with open('assdata.json', mode='r') as file:
    data = json.load(file)   # Load and parse JSON data into a Python dictionary

print("Employees younger than 45:")   # Title for the output

# Loop through each employee in the JSON data
for employee in data['employees']:
    # Check if the employee's age is less than 45
    if employee['age'] < 45:
        # Print the employee's name and age
        print(f"Name: {employee['name']}, Age: {employee['age']}")
```


*   XML


```python
import xml.etree.ElementTree as ET   # Import the ElementTree library for parsing XML data

# Parse the XML file and get the root element
tree = ET.parse('assdata.xml')
root = tree.getroot()

print("Employees younger than 45:")   # Title for the output

# Loop through each <employee> element in the XML
for employee in root.findall('employee'):
    age = int(employee.find('age').text)   # Extract age and convert to integer

    # Check if the employee's age is less than 45
    if age < 45:
        name = employee.find('name').text   # Extract the employee's name
        # Print the employee's name and age
        print(f"Name: {name}, Age: {age}")
```
