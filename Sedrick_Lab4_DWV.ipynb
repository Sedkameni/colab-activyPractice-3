```python
#### - Task 2: Working with SQLite3 and Pandas to_csv().

## 1 -    Install Python SQLite Library.

### 2 - Create a connection. Create a cursor object.
# Create a connection
conn = sqlite3.connect('Rumba.db')

# Create a cursor object
cur = conn.cursor()
```

```python
#### 2.3 - Query database to get employee table and display data.

# Query the database
cur.execute("SELECT * FROM employee")
employees = cur.fetchall()

# Display the rows
print('\nEmployee table content:')
for employee in employees:
    print(employee)
```

```python
#### 2.4 - Creating a join query: Write your Join Select Statement.
#Retrieving data
sql = '''SELECT e.employee_id, e.full_name, \
e.job_role, e.manager_id, d.department_name \
from employee e INNER JOIN department d \
ON e.department_id = d.department_id'''
```

```python
### 2.5 - Query database to get query and load into a Python object.
#Executing the query
cur.execute(sql)

#Fetching 1st row from the table
result = cur.fetchall()
```

```python
#### 2.6 - Load Python object into a dataframe, create column names and display first 5 rows.

# Load result into a Dataframe
join_table = pd.DataFrame(result, \
                          columns=['Employee_Id','Employee_Name',\
                                   'Job_Role','Manager_Id',\
                                   'Department_Name'])
print('\n The top 5 records of Employee and Department tables are:')
print(join_table.head())
```

```python
#### 2.7 - Export the content of dataframe to a CSV file, then close connection.

# Export the result into a CSV file
join_table.to_csv("emp_dep.csv",index=False)

# Close connection
conn.close()
```

```python
### 3.3 - Create a table called ‘manager’ with columns: Id, Manager Name, Title, and Sex.
manager = db.Table('manager', metadata,
db.Column('Id', db.Integer(),primary_key=True),
db.Column('Manager_Name', db.String(255), nullable=False),
db.Column('Title', db.String(255), nullable=False),
db.Column('Sex', db.String(1)))
```

```python
#### Task 4: Working with Pandas to_sql() and SQLAlchemy

### 1 -    Install Python SQLAlchemy Library.
import pandas as pd
import sqlalchemy as db
```

```python
### 4.2 - Create engine, a connection, and instance of the MetaData() function.
#Create an engine
engine = db.create_engine('sqlite:///Rumba.db')

#Create a connection
conn = engine.connect()

# Create an instance of MetaData
metadata = db.MetaData()
```

```python
### 4.3 - Read the CSV file created in the first exercise.
df = pd.read_csv('emp_dep.csv')
df.to_sql(con=engine, name="Emp_Dep", if_exists='replace', index=False)
print('Number of records successfully loaded: ', len(df))
```

```python
### 4.4 - Load the tables, Emp_Dep and manager.
emp_dep = db.Table('Emp_Dep', metadata, autoload_with=engine)
manager = db.Table('manager', metadata, autoload_with=engine)
```

```python
from sqlalchemy import text

# Execute the PRAGMA query to get table info
result = conn.execute(text("PRAGMA table_info(manager)"))

# Print the columns of the manager table
for row in result:
    print(row)
```

```python
### 4.5 - Creating a join query: Write your Join Select Statement with an order clause on Employee_Id.
##Query database to joined table and load into a Python object.
# Create a join query
join_query = db.select(emp_dep.columns.Employee_Id,\
                       emp_dep.columns.Employee_Name,\
                       emp_dep.columns.Job_Role, \
                       emp_dep.columns.Department_Name, \
                       manager.columns.FirstName).\
select_from(emp_dep.join(manager,\
                         emp_dep.columns.Manager_Id \
                         == manager.columns.Id)).\
order_by(emp_dep.columns.Employee_Id)

# Retrieve data
output = conn.execute(join_query)
result = output.fetchall()
```

```python
### 4.6 - Load Python object into a dataframe, display data and close connection.
# Load result into a Dataframe
data = pd.DataFrame(result)
print(data)

# Close connection
conn.close()
```
