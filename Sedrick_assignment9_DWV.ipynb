```python
##### Part 1 - Copy these lines of code into your Python file

import pandas as pd
from sklearn.datasets import load_diabetes

# Load the dataset
diabetes = load_diabetes()
data = pd.DataFrame(data=diabetes.data, columns=diabetes.feature_names)
data['target'] = diabetes.target

# Display first few rows
print(data.head())

# Describe the dataset
print(data.describe())
```

```python
#### Part 2  - Write the right code for the following:

### 2.1 - ARGS and Keyword Args
import matplotlib.pyplot as plt
### - Create a figure of figsize = (10,6)
plt.figure(figsize=(10,6))

### - Plot the first feature ('age') against the 'target' variable using positional arguments. Include a marker, line style, and color keywords.
plt.plot(data['age'], data['target'], marker='o', linestyle=':', color='green')

### - Add a title for your plot and labels for x and y axis.
plt.title('Age vs. Target')
plt.xlabel('Age')
plt.ylabel('Target')

### - Display the plot
plt.show()
```

```python
#### 2.2 - Figures and Subplots

### -  Create a figure with two subplots and figsize=(14,6)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

### - Use a scatter plot for subplot 1 and plot 'bmi' vs 'target'. Include color and marker keywords.
ax1.scatter(data['bmi'], data['target'], color='blue', marker='o', alpha=0.5)

### - Add title and labels for x and y axis to the subplot using set_title, set_xlabel, and set_ylabel.
ax1.set_title('BMI vs. Target')
ax1.set_xlabel('BMI')
ax1.set_ylabel('Target')

### - Use a scatter plot for subplot 2 and plot ' 'bp' vs 'target'. Include different color and different marker from that of subplot 1.
ax2.scatter(data['bp'], data['target'], marker='o', color='red', alpha=0.5, edgecolor='black')

### -Add title and labels for x and y axis to the subplot using set_title, set_xlabel, and set_ylabel.
ax2.set_title('BP vs. Target')
ax2.set_xlabel('BP')
ax2.set_ylabel('Target')

### - Display the plots
plt.show()
```

```python
##### 2.3 - Ticks, Labels, and Legends

#### - Create a figure of figsize = (10,6)
plt.figure(figsize=(10,6))

#### - Use a scatter plot and customize the plot for 's1' vs 'target'. Include color, marker and label keywords.
plt.scatter(data['s1'], data['target'], color='blue', marker='*', label='s1')

##### - Add ticks with fontsize of 10, and labels for x and y axis to the plot.
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.title('s1 vs. Target', fontsize=14, fontweight='bold')
plt.xlabel('s1', fontsize=12)
plt.ylabel('Target', fontsize=12)

#### - Add legend and grid to the plot.
plt.legend(loc=2, fontsize=10)
plt.grid(color='brown', linestyle='--', linewidth=0.5)

#### -  Display the plot.
plt.show()
```

```python
##### 2.4 - Colors, Markers, and Line Styles

### - Create a figure of figsize = (10,6)
plt.figure(figsize=(10,6))

### - Plot 's1', 's2', and 's3' each against 'target' using different colors, markers, and line styles.
plt.plot(data['s1'], data['target'], color='blue', marker='o', linestyle='-', label='s1')
plt.plot(data['s2'], data['target'], color='red', marker='s', linestyle='--', label='s2')
plt.plot(data['s3'], data['target'], color='green', marker='^', linestyle=':', label='s3')

#### - Add title and labels for x and y axis to the plot.
plt.title('Features vs. Target', fontsize=14, fontweight='bold')
plt.xlabel('Features', fontsize=12)
plt.ylabel('Target', fontsize=12)


#### - Add legend to the plot.
plt.legend(loc=1, fontsize=10)

#### - Add grid to the plot.
plt.grid(color='gray', linestyle=':', linewidth=0.7)
### - Display the plot.
plt.show()
```

```python
##### 2.5 - Legends and grid

### - Create a figure of figsize = (10,6).
plt.figure(figsize=(10,6))

### - Plot 's4' vs 'target' and add legends and grids to the plot.
plt.plot(data['s4'], data['target'], color='blue', marker='o', linestyle=':', label='s4')

### - Add grid and legend with loc =2 to the plot.
plt.grid(color='gray', linestyle=':', linewidth=0.7)
plt.legend(loc=2, fontsize=10)

### - Add title and labels for x and y axis to the plot.
plt.title('s4 vs. Target', fontsize=14, fontweight='bold')
plt.xlabel('s4', fontsize=12)
plt.ylabel('Target', fontsize=12)

### - Display the plot.
plt.show()
```

```python
##### 2.6 - Saving Plots as an Image File

### - Create a figure of figsize = (10,6).
plt.figure(figsize=(10,6))

### -  Plot  'age' vs 'target' and add a marker.
plt.plot(data['age'], data['target'], marker='^', linestyle=':', color='green')

### - Add title and labels for x and y axis to the plot.
plt.title('Age vs. Target', fontsize=14, fontweight='bold')
plt.xlabel('Age', fontsize=12)
plt.ylabel('Target', fontsize=12)

### - Save the plot of 'age' vs 'target' as both PNG and PDF files.
plt.savefig('age_vs_target.png')
plt.savefig('age_vs_target.pdf')

#### - Display the plot
plt.show()
```
